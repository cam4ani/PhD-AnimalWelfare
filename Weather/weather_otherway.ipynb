```python
from lxml import html
import pandas as pd
import requests
import datetime as dt
```

```python
url = 'http://www.meteocentrale.ch/en/europe/switzerland/weather-zollikofen/details/S066310/#today'
encoding = 'utf-8'
page = requests.get(url, timeout=5)
content = page.content
tree = html.fromstring(content.decode(encoding))
```

```python
dico = {}
#take info of todays table (table-0)
for type_ in ['Feels like','Relative humidity','Precipitation','Temperature','Avg. wind speed','Dewpoint',
              'Air pressure','Fresh snow','Snow depth']:
    li_meteo = []
    for m in tree.xpath('//table[@id="detail-table-0"]/tr/td[@title="%s"]'%type_):
        li_meteo.append(m.text)
    if len(li_meteo)==3:
        li_meteo = [li_meteo[0],li_meteo[0],li_meteo[0],li_meteo[1],li_meteo[1],li_meteo[2],li_meteo[2]]
    dico[type_] = li_meteo
#add sky
li_meteo = []
for m in tree.xpath('//table[@id="detail-table-0"]/tr/td/center/div[@class="mm_detail_tooltip"]'):
    li_meteo.append(m.text)
dico['sky'] = li_meteo
#add hours
li_hour = []
for i,p in enumerate(tree.xpath('//table[@id="detail-table-0"]/tr/td/div')):
    if not i%2:
        li_hour.append(p.text)
dico['hour'] = li_hour
#current time
t = dt.datetime.now()
dico['date'] = t
df = pd.DataFrame.from_dict(dico)
df
```
