# Download data

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/correctlightschedule_/visual/Treatment&Classs/adaptability'
#df = read.csv(file.path(path_,'df_STAT_RoutineDev_interintra.csv'), header = TRUE, sep = ";")
df = read.csv(file.path(path_,'STATInterIntraDifferences.csv'), header = TRUE, sep = ";")
df$Treatment = as.factor(df$Treatment) 
df$PenID = as.factor(df$PenID)
df$henID1 = as.factor(df$henID1)
df$henID2 = as.factor(df$henID2)
df$inter_intra = as.factor(df$inter_intra)
df$DIB = as.integer(df$DIB) #as factor as within a week its not th etime taht we want to take into acount, but the event
df$factDIB = as.factor(df$DIB)
df$sqrtchi2distance = sqrt(df$chi2distance)
df$sqrtsqrtchi2distance = sqrt(df$sqrtchi2distance)
df$logchi2distance = log(df$chi2distance+1)
print(dim(df))
summary(df)
head(df, 3)
```

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/correctlightschedule_/visual/Treatment&Classs/adaptability'
df_intra = read.csv(file.path(path_,'df_STAT_RoutineDev_intraRearingFeatures.csv'), header = TRUE, sep = ";")
df_intra$PenID = as.factor(df_intra$PenID)
df_intra$HenID = as.factor(df_intra$HenID)
df_intra$RPenID = as.factor(df_intra$RPenID)
df_intra$Treatment = as.factor(df_intra$Treatment)
df_intra$DIB = as.integer(df_intra$DIB)
df_intra$factDIB = as.factor(df_intra$DIB)
df_intra$HenID1_CLASS = as.factor(df_intra$HenID1_CLASS)
df_intra$sqrtchi2distance = sqrt(df_intra$chi2distance)
df_intra$sqrtsqrtchi2distance = sqrt(df_intra$sqrtchi2distance)
df_intra$logchi2distance = log(df_intra$chi2distance+1)
print(dim(df_intra))
summary(df_intra)
head(df_intra, 3)
```

# Rearing features association

```r
df_intra$juin_weight.scaled = rescale(df_intra$juin_weight)
df_intra$percentage_of_gain_weight_rearingbarn.scaled = rescale(df_intra$percentage_of_gain_weight_rearingbarn)
df_intra$DIB.scaled = rescale(df_intra$DIB)
df_intra$DIB.scaled = rescale(df_intra$DIB)
df_intra$max_Total_number_transition.scaled = rescale(df_intra$max_Total_number_transition)
df_intra$min_Total_number_transition.scaled = rescale(df_intra$min_Total_number_transition)
```

```r
head(df_intra,5)
```

```r
#modelling
fit = lmerTest::lmer(sqrtchi2distance ~ max_Total_number_transition + min_Total_number_transition + HenID1_CLASS*DIB + RPenID*DIB + (1|PenID/HenID), 
           data=df_intra)
#PenID
#capture.output(summary(fit), file=file.path(path_,paste(name,'.txt', sep='')), append=TRUE)
summary(fit)
#tests the hypothesis that the random effect variation is significant versus the null that there is no between group-level 
#variation
rand(fit)
confint(fit)
anova(fit)

#Assumptions
######linearity of predictors: plot residual vs observed
plot(resid(fit), df_intra$sqrtchi2distance)

###### normally distributed residuals
#qqmath(fit, id=0.05)
qqnorm(resid(fit), main = "Residuals")
qqline(resid(fit))
hist(resid(fit))
###### normally of random effects
qqmath(ranef(fit,condVar = TRUE))

######check homogeneity of variance (residuals has constant variance)
#variance of the residuals is equal across groups. Even spread around the centered line?
plot(fit)

###### check if the residuals are independent. This can be checked by plotting residuals against covariates
plot(resid(fit),df_intra$percentage_of_gain_weight_rearingbarn.scaled)
plot(resid(fit),df_intra$juin_weight.scaled)
```

```r
#marginal effecct. TO READ 
#https://cran.r-project.org/web/packages/sjPlot/vignettes/plot_interactions.html 
#and verify if apply to these models
#plot_model(fit, type = "pred", terms = c("DIB", "EarlyGroup"))
#more general but exact same output in this case:
plot_model(fit, type = "int")
```

```r
df_NOMEXP = df[ which(df$is_only_other=='True'), ]
df_NOMEXP$hoDIB = df_NOMEXP$DIB**2
df_NOMEXP$sqrtmin_Total_number_transition = sqrt(df_NOMEXP$min_Total_number_transition)
df_NOMEXP$sqrtmax_Total_number_transition = sqrt(df_NOMEXP$max_Total_number_transition)
df_NOMEXP$min_Total_number_transition2 = df_NOMEXP$min_Total_number_transition**2
df_NOMEXP$max_Total_number_transition2 = df_NOMEXP$max_Total_number_transition**2
df_NOMEXP$max_Total_number_transition2 = scale(df_NOMEXP$max_Total_number_transition2, center = FALSE, 
                                               scale = max(df_NOMEXP$max_Total_number_transition2, na.rm = TRUE)/100)
df_NOMEXP$min_Total_number_transition2 = scale(df_NOMEXP$min_Total_number_transition2, center = FALSE, 
                                               scale = max(df_NOMEXP$min_Total_number_transition2, na.rm = TRUE)/100)
dim(df_NOMEXP)
summary(df_NOMEXP)
tail(df_NOMEXP,3)
```

```r
#choosing data: ONLY inter to compare treatment FINAL 1
#with only non mexp
df_NOMEXPINTER = df_NOMEXP[ which(df_NOMEXP$inter_intra=='inter'), ]
dim(df_NOMEXPINTER)
tail(df_NOMEXPINTER,3)

#modelling
fit = lmerTest::lmer(sqrtchi2distance ~ max_Total_number_transition + min_Total_number_transition + DIB + Treatment*DIB + (1|PenID/henID1), 
           data=df_NOMEXPINTER)
#capture.output(summary(fit), file=file.path(path_,paste(name,'.txt', sep='')), append=TRUE)
summary(fit)
#tests the hypothesis that the random effect variation is significant versus the null that there is no between group-level 
#variation
rand(fit)
confint(fit)

#Assumptions
######linearity of predictors: plot residual vs observed
plot(resid(fit), df_NOMEXPINTER$sqrtchi2distance)

###### normally distributed residuals
#qqmath(fit, id=0.05)
qqnorm(resid(fit), main = "Residuals")
qqline(resid(fit))
hist(resid(fit))
###### normally of random effects
qqmath(ranef(fit,condVar = TRUE))

######check homogeneity of variance (residuals has constant variance)
#variance of the residuals is equal across groups. Even spread around the centered line?
plot(fit)

###### check if the residuals are independent. This can be checked by plotting residuals against covariates - especially time-varying or spatial covariates. There should not be any systematic pattern
plot(resid(fit),df_NOMEXPINTER$DIB)
plot(resid(fit),df_NOMEXPINTER$max_Total_number_transition)
plot(resid(fit),df_NOMEXPINTER$min_Total_number_transition)
```
