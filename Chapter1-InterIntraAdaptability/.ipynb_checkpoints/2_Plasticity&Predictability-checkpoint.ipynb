# Download data

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/ALLDATA_'
path_adapt = file.path(path_,'Adaptability')
path_model = file.path(path_adapt,'repeatability_model')
df = read.csv(file.path(path_adapt,'df_MVT_4individuality.csv'), header = TRUE, sep = ";")
df$HenID = as.factor(df$HenID)   
df$PenID = as.factor(df$PenID) 
df$WIB = as.factor(df$WIB) 
df$TrackingSystemID = as.factor(df$TrackingSystemID) 
#mean-centering of the environmental variable so that intercepts reflect average values for the population and individuals
df$cDIB = as.integer(df$DIB) 
df$cDIBfact = as.factor(df$cDIB) 
df$RearingPenID = as.factor(df$RearingPenID) 
df$InitialWeight_kg = df$InitialWeight/1000
df$CLASS = as.factor(df$CLASS) 
df$Treatment = as.factor(df$Treatment) 
print(dim(df))
summary(df)
```

# PCA

```r
#define var and scale them 
li_pca = c('perc_duration_5_Zone', 'perc_duration_4_Zone','perc_duration_3_Zone', 'perc_duration_2_Zone','perc_1_Zone_while_WG_open',
          'nbr_stays_2_Zone_perh', 'nbr_stays_4_Zone_perh','nbr_stays_3_Zone_perh', 'nbr_stays_5_Zone_perh','nbr_stays_1_Zone_perh',
          'in_WG_15mnAfterOpening','distribution_entropy', 'SleepingHeight','vertical_travel_distance_perh')
df[li_pca] = scale(df[li_pca])
```

```r
#frst obs per week per individual
df_pca = df[df$tobeusedPCA==1,][li_pca]
head(df_pca,3)
```

```r
#https://www.rdocumentation.org/packages/psych/versions/2.1.9/topics/principal
res = psych::principal(df_pca, rotate="none", nfactors=3, scores=TRUE) #not varimax (we want PC, and only one)
res
```

```r
#for validity compare with the one run on full dataset
res_validity = psych::principal(df[li_pca], rotate="none", nfactors=3, scores=TRUE) #not varimax (we want PC, and only one)
res_validity
```

```r
#documentation: The beta weights to find the principal components from the data
res$weights
```

```r
#eigenvalues of all components --> first one >1
res$values
```

```r
#https://www.rdocumentation.org/packages/psych/versions/2.1.6/topics/predict.psych
df_result = data.frame(predict(res, df[li_pca]))
df$PC1 = df_result$PC1
head(df,3)
```

```r
#documentation: https://www.rdocumentation.org/packages/psych/versions/2.0.9/topics/biplot.psych
biplot(res)
```

# PenID/Class/RearingPenID - check their effects

```r
modeltestinit = lmer(PC1 ~ 1 + (1|HenID), REML = FALSE, data = df)
summary(modeltestinit)
r.squaredGLMM(modeltestinit)

modeltest1 = lmer(PC1 ~ 1 + (1|PenID/HenID), REML = FALSE, data = df)
summary(modeltest1)
r.squaredGLMM(modeltest1)

modeltest2 = lmer(PC1 ~ CLASS + (1|HenID), REML = FALSE, data = df)
summary(modeltest2)
r.squaredGLMM(modeltest2)

modeltest3 = lmer(PC1 ~ RearingPenID + (1|HenID), REML = FALSE, data = df)
summary(modeltest3)
r.squaredGLMM(modeltest3)

summary(modeltestinit)$logLik
summary(modeltest1)$logLik
summary(modeltest2)$logLik
summary(modeltest3)$logLik

chi2 = 2*(summary(modeltest1)$logLik -summary(modeltestinit)$logLik)
1-pchisq(chi2,1) # chi2, df

chi2 = 2*(summary(modeltest2)$logLik -summary(modeltestinit)$logLik)
1-pchisq(chi2,2) # chi2, df

chi2 = 2*(summary(modeltest3)$logLik -summary(modeltestinit)$logLik)
1-pchisq(chi2,3) # chi2, df

AIC(modeltestinit, modeltest1, modeltest2, modeltest3)
```

# Behavioural type

```r
modelBT = lmer(PC1 ~ cDIB + Treatment + temperature_C_avg + InitialWeight_kg + (1|HenID), REML = FALSE, data = df)
summary(modelBT)
r.squaredGLMM(modelBT)
```

```r
###### normally distributed residuals
qqnorm(resid(modelBT))
qqline(resid(modelBT))
hist(resid(modelBT))
######check homogeneity of variance (residuals has constant variance)
plot(modelBT)
```

```r
#repeatabilities
print(VarCorr(modelBT),comp =c("Variance","Std.Dev."))
rep = VarCorr(modelBT)$"HenID"[1]/
(VarCorr(modelBT)$"HenID"[1]+
 attr(VarCorr(modelBT), "sc")^2) # equal to the variance of residual
#alternatively: 0.39815  /(0.39815  +0.47685)
rep
```

# Plasticity

```r
#random interecepts and slopes over DIB for each animal (interactio between DIB and henID as a random effect)
modelPL = lmer(PC1 ~ cDIB + Treatment + temperature_C_avg + InitialWeight_kg + (1 + cDIB|HenID), 
                REML = FALSE, data = df,
              control = lmerControl(optimizer ='optimx', optCtrl=list(method='nlminb')))
summary(modelPL)
r.squaredGLMM(modelPL)
```

```r
###### normally distributed residuals
qqnorm(resid(modelPL))
qqline(resid(modelPL))
hist(resid(modelPL))
######check homogeneity of variance (residuals has constant variance)
plot(modelPL)
```

```r
#test whether a nonlinear random slope over DIB improve model fit
modelPL2 = lmer(PC1 ~ poly(cDIB, 2) + Treatment + temperature_C_avg + InitialWeight_kg + (1+poly(cDIB, 2)|HenID), 
                REML = FALSE, data = df)
#control = lmerControl(optimizer ='optimx', optCtrl=list(method='nlminb'))
summary(modelPL2)
r.squaredGLMM(modelPL2)
```

```r
###### normally distributed residuals
qqnorm(resid(modelPL2))
qqline(resid(modelPL2))
hist(resid(modelPL2))
######check homogeneity of variance (residuals has constant variance)
plot(modelPL2)
```

```r
#compare the log-likelihoods of the two models and perform a likelihood ratio test (LRT).
#Here, we take 2 ×the log-likelihood values from the model summaries to determine a 𝜒+value, 
#and then determine the p-value of the LRT from the 𝜒+value and the difference of the degrees of freedom(df) between the two models using the function 1-pchisq(chi2, df)
#CHECK df difference between models (by looking at the df within the models being compared)
summary(modelPL)$logLik
summary(modelBT)$logLik
chi2 = 2*(summary(modelPL)$logLik -summary(modelBT)$logLik)
1-pchisq(chi2,2) # chi2, df
```

```r
#compare the log-likelihoods of the two models and perform a likelihood ratio test (LRT).
#Here, we take 2 ×the log-likelihood values from the model summaries to determine a 𝜒+value, 
#and then determine the p-value of the LRT from the 𝜒+value and the difference of the degrees of freedom(df) between the two models using the function 1-pchisq(chi2, df)
#CHECK df difference between models (by looking at the df within the models being compared)
summary(modelPL2)$logLik
summary(modelPL)$logLik
chi2 = 2*(summary(modelPL2)$logLik -summary(modelPL)$logLik)
1-pchisq(chi2,4) # chi2, df
```

```r
AIC(modelPL, modelBT, modelPL2)
#--> the more complex model fits better, sugegsting that individuals adapt their mvt differently over the first month in a laying barn
```

# Predictibility

```r
df$temperature_C_avg_scale = scale(df$temperature_C_avg, scale=TRUE)
df$InitialWeight_kg_scale = scale(df$InitialWeight_kg, scale=TRUE)
df$cDIB_scale = scale(df$cDIB, scale=TRUE)
head(df,3)
```

```r
#RearingPenID, Treatment, CLASS
```

```r
#Get information on all parameters (and parameter classes) for which priors may be specified including default priors:
get_prior(formula=PC1~ cDIB + temperature_C_avg + (1+cDIB|HenID), data=df, family = gaussian())
#family: A description of the response distribution and link function to be used in the model
#If in doubt, run the analysis with multiple prior settings (i.e. “multiverse analysis”) and inquire further if any of your 
#inferences change based on prior choice.
#Prior predictive checks are a way to make sure your priors for model parameters are consistent with your beliefs about 
#plausible outcomes.
#Flat and super-vague priors are not usually recommended and some thought should included to have at least weakly informative 
#priors.
#https://github.com/stan-dev/stan/wiki/Prior-Choice-Recommendations
#https://rss.onlinelibrary.wiley.com/doi/10.1111/rssa.12378
```

```r
### bulk-ESS: We recommend requiring that the bulk-ESS is greater than 100 times the number of chains. For example, when running 
#4 chains, this corresponds to having a rank-normalized effective sample size of at least 400.
### R-hat: We recommend running at least four chains by default and only using the sample if R-hat is less than 1.01.
#more info: #https://mc-stan.org/misc/warnings.html#bulk-ess%22
double_model = bf(PC1~ cDIB_scale + temperature_C_avg_scale + (1+cDIB_scale|HenID), sigma~(1|HenID))
modelPred = brm(double_model, data=df, iter=12000, inits="random", seed=12345, control = list(max_treedepth = 15), cores=my.cores,
             chains=6, thin=2)
             #prior = c(set_prior("student_t(1, 0.01, 0.001)", coef = "temperature_C_avg_scale"),
             #         set_prior("student_t(1, 0.01, 0.001)", coef = "cDIB_scale")))
#modelPred = add_criterion(modelPred, "waic")
summary(modelPred)
pairs(modelPred)
#We can also compute the “coefficient of variation in predictability” (CVP) a standardized population-levelmeasure of the 
#degree of variation in predictability among all our study individuals.CVPcan be used tocompare the degree of among-individual 
#variation in predictability across studies and traits and we thereforerecommend authors always to report it
log.norm.res = exp(posterior_samples(modelPred)$"sd_HenID__sigma_Intercept"^2)
CVP = sqrt(log.norm.res-1)
mean(CVP);HPDinterval(as.mcmc(CVP),0.95)
df_pred = posterior_samples(modelPred)
write.csv(df_pred, file.path(path_,'Adaptability','IndividualPredictability.csv'), row.names = FALSE)
prior_summary(modelPred)
```
