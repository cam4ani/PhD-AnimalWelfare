```python
#basic package
import numpy as np
import pandas as pd
import datetime as dt
import os
import math
#plot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'G:\VPHI\Welfare\2- Research Projects\OFHE2.OriginsE2\Epigenetic\sent_Fabio_18-11-2022'
```

## download data

```python
#epi info
df_epi = pd.read_csv(os.path.join(path, 'epi_recordswithHenID.csv'), sep=',', parse_dates=['Date'], dayfirst=True)
df_epi.rename(columns={'LegringID':'FocalLegringName'}, inplace=True)
dico_Rpen_tr = {1:'OFH',2:'TRAN',3:'OFH',4:'TRAN'}
df_epi['Treatment'] = df_epi['Rpen'].map(lambda x: dico_Rpen_tr[x])
#except first timestamp, no treatement yet! we used three pens for hatching
df_epi.loc[df_epi['TimestampID']==0,'Treatment'] = 'OFH-eggs'
#add housing info
df_epi['Housing'] = df_epi['TimestampID'].map(lambda x: 'RearingBarn1' if x<=6 else 'LayingBarn4')
#add henID when in rearing
df_epi['HenID'] = df_epi['HenID'].map(lambda x: 'hen_'+str(int(x)) if not math.isnan(x) else x)
df_epi.loc[df_epi['Housing']=='RearingBarn1','HenID'] = 'NotExisting'
print(df_epi.shape)
display(df_epi.head(3))
```

```python
#dowloand focal info (tags changes, legring change, weight etc)
df_Focal = pd.read_csv(os.path.join(path, 'ALLDATA_df_FOCALBIRDS.csv'), encoding='latin', sep=';', 
                       parse_dates=['InitialStartDate'], dayfirst=True)
print(df_Focal.shape)
display(df_Focal.head(3))
```

# Merge info for birds in the laying phase

```python
df = pd.merge(df_epi, df_Focal, on=['HenID','Treatment'], how='left')
print(df_epi.shape, df_Focal.shape, df.shape)
#blind Fabio
dico_ = {'OFH':'elephant','TRAN':'crocodile','OFH-eggs':'Eggs'}
df['BlindedTreatment'] = df['Treatment'].map(lambda x: dico_[x])
#save
df.to_csv(os.path.join(path, 'EpiAllInfo.csv'),sep=';',index=False)
df.to_csv(os.path.join(path, 'EpiAllInfo_comma.csv'),sep=',',index=False)
display(df.tail(3))
```
