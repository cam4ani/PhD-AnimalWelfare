```python
li_type = ['temperature', 'humidity', 'wind speed', 'pressure', 'Precipitation amount']
li_type = ['temperature', 'Precipitation amount']
```

```python
mi = dt.datetime(2019,9,1)
#ma = dt.datetime(2019,9,2)
ma = dt.datetime(2020,7,30)
df = pd.DataFrame(columns=['date', 'temperature', 'humidity', 'wind speed', 'pressure', 'Precipitation amount'])
if len(glob.glob('weather.csv'))==1:
    df = pd.read_csv('weather.csv', sep=';', parse_dates=['date'])
    print('We already have %d dates of weather'%df.shape[0])
    mi = max(mi, max(df['date'].tolist()))

Daterange = pd.date_range(start = mi, end = ma, freq = 'H')    
if (len(Daterange)*2)>=2000:
    print('WARNING: to many calls to be completed these queries!')
print('You need %d new queries to complete this date interval'%(len(Daterange)*len(li_type)))
```

```python
#not same as : https://weather.plus/observations-weather-station.php?dest=Bern-Zollikofen&id=066310&type=synop&day=20200713, our
#results are less consistent :( 
#ATTENTION: monthly api calls limited to 2'000
#open if exist
li_df = []
try:
    for d in tqdm.tqdm(Daterange):
        dico = {'date':d}
        for type_ in li_type:
            query = '%s in Zollikofen on %s'%(type_,str(d))
            res = client.query(query)
            for t in res.results:
                dico[type_] = t.text
        li_df.append(dico)
except Exception as e:
    print('ERROR:', e)
df_ = pd.DataFrame(li_df)
df = pd.concat([df,df_])
df.to_csv('weather.csv', index=False, sep=';')
print(df.shape)
display(df.head(3))
```
