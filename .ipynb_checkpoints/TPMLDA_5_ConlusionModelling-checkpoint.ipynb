```python
#'topic_modelling_and_human_var','human_var_only'
li_path_save = glob.glob(os.path.join(path_extracted_data,'visual','predict_KBF', '*','*', '*_coefs_*.csv'))
li_path_save = [x for x in li_path_save if 'old' not in x]
li_dfcoef = []
li_model = []
for path_save in li_path_save:
    
    #extract info
    Sess2keep = path_save.split('---sess_')[1].split('\\')[0].split('_')
    Sess2keep = [int(i) for i in Sess2keep]
    type_hybrid = path_save.split('\\')[-3].split('---')[0]
    type_ = path_save.split('\\')[-3].split('---')[1]
    comment = path_save.split('\\')[-3].split('---')[2]
    lda_type = path_save.split('\\')[-2]
    #random_seed = path_save.split('\\')[-1].split('_')[0]
    
    #coef
    df_coefs = pd.read_csv(path_save, sep=';', index_col=0)
    df_coefs['sessions'] = '_'.join([str(i) for i in Sess2keep])
    df_coefs['type_hybrid'] = type_hybrid
    df_coefs['type_'] = type_
    df_coefs['comment'] = comment  
    df_coefs['lda_type'] = lda_type
    df_coefs['model_path'] = path_save
    df_coefs['output'] = path_save.split('\\')[-1].split('coefs_')[-1].split('.csv')[0]
    li_dfcoef.append(df_coefs)

    #model
    df_model = pd.read_csv(path_save.replace('coefs','model_output'), sep=';', index_col=0)
    df_model['sessions'] = '_'.join([str(i) for i in Sess2keep])
    df_model['type_hybrid'] = type_hybrid
    df_model['type_'] = type_
    df_model['comment'] = comment
    df_model['lda_type'] = lda_type
    df_model['model_path'] = path_save
    df_model['output'] = path_save.split('\\')[-1].split('coefs_')[-1].split('.csv')[0]
    li_model.append(df_model)
    
df_coefs = pd.concat(li_dfcoef)
print(df_coefs.shape)
df_coefs['first_session_considered'] = df_coefs['sessions'].map(lambda x: int(x.split('_')[0]))
df_coefs['nbr_sessions'] = df_coefs['sessions'].map(lambda x: len(x.split('_')))
li = ['estimate','std.error','statistic','p.value']
df_coefs[li] = df_coefs[li].applymap(lambda x: float(str(x).replace(',','.')))
display(df_coefs.head(3))

df_model = pd.concat(li_model)
df_model['first_session_considered'] = df_model['sessions'].map(lambda x: int(x.split('_')[0]))
df_model['nbr_sessions'] = df_model['sessions'].map(lambda x: len(x.split('_')))
li = ['r.squared', 'adj.r.squared', 'sigma', 'statistic', 'p.value','logLik','AIC','BIC','deviance']
df_model[li] = df_model[li].applymap(lambda x: float(str(x).replace(',','.')))
print(df_model.shape)
display(df_model.head(3))
```

```python
df_ = df_coefs.groupby(['model_path'])['term','estimate','std.error','statistic','p.value'].agg(lambda x: str(list(x))).reset_index()
print(df_model.shape, df_.shape)
df_model = pd.merge(df_model, df_, on='model_path', how='outer')
df_model['nbr_param_selected'] = df_model['term'].map(lambda x: len(eval(x)))
df_model['has_maxestimvalue'] = df_model['model_path'].map(lambda x: x.split('_')[-2].strip().isdigit())
df_model['nbr_estimate_max_constraint'] = df_model.apply(lambda x: int(x['model_path'].split('_')[-2]) if x['has_maxestimvalue'] else np.nan,
                                                         axis=1)
print(df_model.shape)
df_model.head(3)
```
