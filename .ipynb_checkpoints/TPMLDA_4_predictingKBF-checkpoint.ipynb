```r
li_path = Sys.glob(file.path("D:/vm_exchange", 'AVIFORUM','data','extracted_info_mobility_VF','visual','predict_KBF',
                             '*', #varyinglength
                             '*','df_IDS_onall.csv'))
print(length(li_path))

li_var_cont = c('ID0','ID1','ID2','ID3','ID4','ID5','ID6','ID7')
x = 'max_severity' #ratio_HealingAndNonhealing_round, max_positive_rate_in_severity, value_of_first_positive_rate_in_severity
#x = 'maxsevGap0_divided_maxsev'
#x = 'X..Width....mean..'
#x = 'X..Mass....mean..'
#x = 'AvgSeverity_gap1.AvgSeverity_gap0'
#4 CHANGE IN THE CODE!!!
nbr_2keep_max = 8
#fix for all run the same random seed
li_r = sample(2000:20000, 20, replace=F)
for (path_ in li_path){
    print(path_)
    path_save_ = strsplit(path_, '/df_IDS_onall.csv', fixed = FALSE, perl = FALSE, useBytes = FALSE)[[1]]
    df = read.csv(path_, header = TRUE, sep = ";") #na.strings=c("","NA")
    df_ = df[,c(li_var_cont, 'max_severity')] 
    #normalize
    #df_[,li_var_cont] = scale(df_[,li_var_cont])
    #print(summary(df_))
    #print(dim(df_))
    
    #Other alternatives are the penalized regression (ridge and lasso regression) (Chapter @ref(penalized-regression)) 
    #and the principal components-based regression methods (PCR and PLS) (Chapter @ref(pcr-and-pls-regression)).
    k = 0
    for (r in li_r){
        set.seed(r)
        # Set up repeated k-fold cross-validation
        train.control <- trainControl(method = "cv", number=3) #above three for 900 obs is not possible
        # Train the model
        #method: "leapBackward","leapForward", "leapSeq" (to fit linear regression with stepwise selection)
        #RMSE and MAE are two different metrics measuring the prediction error of each model. The lower the better
        step.model <- train(max_severity ~., data=df_,
                            method = "leapSeq", 
                            tuneGrid = data.frame(nvmax=1:8),
                            trControl = train.control, na.action = na.exclude) #na.exclude na.omit
        step.model$results
        nv = step.model$bestTune
        #print(nv)
        #print(summary(step.model$finalModel))

        #for all selection, it select the row and take the var which are True
        df_sum = as.data.frame(summary(step.model$finalModel)$which)
        li_2keep = unlist(apply(df_sum, 1, function(x) list(names(which(x==TRUE)))), recursive=FALSE)[[toString(min(nbr_2keep_max,
                                                                                                                    nv[[1]]))]]
        li_2keep = li_2keep[c(-1)] #remove intercept
        #print(li_2keep)
        model = lm(max_severity~., data=df_[c(li_2keep,c('max_severity'))])
        #print(summary(model))
                                
        #remove outliers with cooks measure (multivariate outliers removal)
        #In general use, those observations that have a cookâ€™s distance greater than 4 times the mean may be classified as influential. 
        #cooksd <- cooks.distance(model)
        #plot(cooksd, pch="*", cex=2, main="Influential Obs by Cooks distance")  # plot cook's distance
        #abline(h = 4*mean(cooksd, na.rm=T), col="red")  # add cutoff line
        #text(x=1:length(cooksd)+1, y=cooksd, labels=ifelse(cooksd>4*mean(cooksd, na.rm=T),names(cooksd),""), col="red")  # add labels
        #influential <- as.numeric(names(cooksd)[(cooksd > *mean(cooksd, na.rm=T))])  # influential row numbers
        #print(dim(df_[influential, ])) # nbr of influential observations.
        #df_ = df_[-influential, ]
        #model = lm(max_severity~., data=df_[c(li_2keep,c('max_severity'))])

        #save the model output
        df_model = tidy(model)
        df_model$random_seed = r
        df_coeff = glance(model)
        df_coeff$random_seed = r
        write.csv2(df_model, file.path(path_save_, paste0(toString(r),'_',toString(nbr_2keep_max),'_coefs_',x,'.csv')))
        write.csv2(df_coeff, file.path(path_save_, paste0(toString(r),'_',toString(nbr_2keep_max),'_model_output_',x,'.csv')))

        #a plot of residuals against fitted values, a Scale-Location plot of sqrt(| residuals |) against fitted values, a Normal 
        #Q-Q plot, a plot of Cook's distances versus row labels, a plot of residuals against leverages, and a plot of Cook's distances
        #against leverage/(1-leverage). By default, the first three and 5 are provided.
        #https://stat.ethz.ch/R-manual/R-devel/library/stats/html/plot.lm.html
        #id.n: number of points to be labelled in each plot, starting with the most extreme. (Null, 15,...)
        #par(mfrow = c(2, 2), oma = c(0, 0, 2, 0))
        #plot(model, id.n=15, labels.id=NULL, which=1:6) 
        #residual vs fitted: mean of zero, random, cst variance
        k = k+1
        #plot only three plot per model
        if (k<4){
                p = file.path(path_save_, 'visual')
                dir.create(p, showWarnings = FALSE)
                setwd(p)
                jpeg(file.path(p, paste0(toString(r),'_',toString(nbr_2keep_max),'_diagnostic_plot_',x,'.jpg')))
                lindia::gg_diagnose(model)
                dev.off()
            }
        }
}
```

```r
#remove outliers with cooks measure (multivariate outliers removal)
#In general use, those observations that have a cookâ€™s distance greater than 4 times the mean may be classified as influential. 
cooksd <- cooks.distance(model)
plot(cooksd, pch="*", cex=2, main="Influential Obs by Cooks distance")  # plot cook's distance
abline(h = 4*mean(cooksd, na.rm=T), col="red")  # add cutoff line
text(x=1:length(cooksd)+1, y=cooksd, labels=ifelse(cooksd>4*mean(cooksd, na.rm=T),names(cooksd),""), col="red")  # add labels
influential <- as.numeric(names(cooksd)[(cooksd > 4*mean(cooksd, na.rm=T))])  # influential row numbers
print(dim(df_[-influential, ]))
print(dim(df_))
print(dim(df_[influential, ])) # influential observations.
df_ = df_[-influential, ]
```
