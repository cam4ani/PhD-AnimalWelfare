```r
#no gap, but to choose distribution
#ziformula = ~0 by default
#The basic glmmTMB fit — a zero-inflated Poisson model with a single zeroinflation parameter applying to all observations (ziformula~1). (Excluding
#zero-inflation is glmmTMB’s default: to exclude it explicitly, use ziformula~0.)
#use offset(log to model rate. as explain in https://stats.stackexchange.com/questions/11182/when-to-use-an-offset-in-a-poisson-regression)
NB_mod_P = glmmTMB(duration_5_Zone_min ~ poly(severity_div10, degree=2) + poly(FeatherDamage_div10, degree=2) +offset(log(time_wg_open_h))+ scale(age) + scale(temperature_C_avg) + Treatment + (1|PenID/HenID)+(1|date_HA/date),
                data = df_, ziformula=~0,
                family="poisson")
NB_mod_NB2 = glmmTMB(duration_5_Zone_min ~ poly(severity_div10, degree=2) + poly(FeatherDamage_div10, degree=2) +offset(log(time_wg_open_h))+ scale(age) + scale(temperature_C_avg)+ Treatment + (1|PenID/HenID)+(1|date_HA/date),
                data = df_, ziformula=~0,
                family="nbinom2")
NB_mod_NB1 = glmmTMB(duration_5_Zone_min ~ poly(severity_div10, degree=2) + poly(FeatherDamage_div10, degree=2) +offset(log(time_wg_open_h))+ scale(age) + scale(temperature_C_avg)+ Treatment + (1|PenID/HenID)+(1|date_HA/date),
                data = df_, ziformula=~0,
                family="nbinom1")
AICtab(NB_mod_P, NB_mod_NB2, NB_mod_NB1)
AIC(NB_mod_P, NB_mod_NB2, NB_mod_NB1)
# much lower, much better
#we can Relax the assumption that the duration_1_Zone_min is strictly proportional to time_wg_open_h and use time_wg_open_h instead of offset(log(time_wg_open_h))
```

```r
summary(NB_mod_NB1)
sim_res = simulateResiduals(NB_mod_NB1)
plot(sim_res)
```

```r
plot(predictorEffect("FeatherDamage_div10", NB_mod_NB1), type = "response")
```

```r
#as.data.frame(Effect('severity_div10', mod))
df_res_pred = as.data.frame(Effect('FeatherDamage_div10', NB_mod_NB1))
df_res_pred
ggplot(data=df_, aes(FeatherDamage_div10, duration_5_Zone_min)) + geom_point(alpha = 0.2) + geom_line(data=df_res_pred, col='blue', aes(FeatherDamage_div10, fit))#+geom_ribbon(data=df_res_pred, aes(ymin=lower_s, ymax=upper_s),alpha = 0.1, color='blue')
```

```r
#MODEL FIT
#glmer assumption with dharma: https://cran.r-project.org/web/packages/DHARMa/vignettes/DHARMa.html
mod = lmerTest::lmer(sqrt(duration_5_Zone_min) ~ severity_div10 + FeatherDamage_div10 + scale(age) + scale(temperature_C_avg) + Treatment + (1|PenID/HenID)+(1|date_HA/date), 
           data=df_, control=lmerControl(optimizer="bobyqa", optCtrl=list(maxfun=1e+05)))
summary(mod)
#Assumptions
######linearity of predictors: plot residual vs observed
plot(resid(mod), sqrt(df_$duration_5_Zone_min))

###### normally distributed residuals
#qqmath(mod, id=0.05)
qqnorm(resid(mod), main = "Residuals")
qqline(resid(mod))
hist(resid(mod))
###### normally of random effects
qqmath(ranef(mod,condVar = TRUE))

######check homogeneity of variance (residuals has constant variance)
#variance of the residuals is equal across groups. Even spread around the centered line?
plot(mod)

###### check if the residuals are independent. This can be checked by plotting residuals against covariates
plot(resid(mod),df_$FeatherDamage_div10)
plot(resid(mod),df_$severity_div10)
plot(resid(mod),df_$age)

#PLOT
#effect_plot: https://www.rdocumentation.org/packages/jtools/versions/2.1.4/topics/effect_plot
#"This function supports nonlinear and generalized linear models and by default will plot them on their original scale
#While mixed effects models from lme4 are supported, only the fixed effects are plotted. lme4 does not provide CI, so 
#they are not supported with this function either.
#Note: to use transformed predictors, e.g., log(x), or polynomials, e.g., poly(x, 2), provide the raw variable name (x) to the 
#pred = argument. You will need to input the data frame used to fit the model with the data = argument."
#partial.residuals= TRUE if instead of plotting the observed data, we want the partial residuals (controlling for the effects of variables besides pred).
effect_plot(mod, pred=severity_div10, interval=TRUE, plot.points=TRUE, partial.residuals=FALSE, data=df_, point.alpha=0.2)+xlab("severity")+ylab('duration_5_Zone_min')
effect_plot(mod, pred=severity_div10, interval=TRUE, plot.points=TRUE, partial.residuals=TRUE, data=df_, point.alpha=0.3)+xlab("severity")+ylab('duration_5_Zone_min')

#ggsave(file.path(path_,'Severity_Predictability_mean.png'), width=4, height=4, bg = "transparent")
effect_plot(mod, pred=FeatherDamage_div10, interval=TRUE, plot.points=TRUE, partial.residuals=FALSE, data=df_, point.alpha=0.2)+xlab("featherdamage")+ylab('duration_5_Zone_min')
effect_plot(mod, pred=FeatherDamage_div10, interval=TRUE, plot.points=TRUE, partial.residuals=TRUE, data=df_, point.alpha=0.2)+xlab("featherdamage")+ylab('duration_5_Zone_min')
```
