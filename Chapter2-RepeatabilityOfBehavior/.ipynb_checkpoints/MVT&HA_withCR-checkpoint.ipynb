```r
library(parallel)
my.cores = detectCores()
```

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/ALLDATA_/chapter2'
df = read.csv(file.path(path_,'df_daily_KBF_ALL.csv'), header = TRUE, sep = ",")
df$HenID = as.factor(df$HenID)   
df$PenID = as.factor(df$PenID) 
df$date = as.factor(df$date) 
df$ExID = as.factor(df$ExID) 
print(dim(df))
summary(df)
df_ = subset(df, !is.na(WentinWG))
dim(df_)

df_ = subset(df_, !is.na(severity_div10))
dim(df_)
```

```r
#s(): "Functions used in definition of smooth terms within a model formulas. The function does not evaluate a (spline) smooth - 
#it exists purely to help set up a model using spline based smooths."
double_model = bf(sqrt(duration_5_Zone_min) ~ severity_div10 + ExID + WOA + (1|HenID) + (1|date),
                  sigma ~ severity_div10 + WOA + ExID + (1|HenID)+ (1|date))
modelPred = brm(double_model, data=df, iter=10000, inits="random", seed=12345, control = list(max_treedepth=15), 
                cores=my.cores, chains=3, thin=10)
#modelPred = add_criterion(modelPred, criterion=c('waic','bayes_R2','loo'), file=file.path(path_,'BRMS_mact_HA_verttraveldistance'))
summary(modelPred)
conditional_smooths(modelPred, stype = "raster")
conditional_effects(modelPred)
```

```r
#Christina
ZI_mod_TRB2P = glmmTMB(duration_1_Zone_min ~ poly(severity_div10, degree=2)+ ExID +scale(WOA) + (1|HenID)+(1|date),
                data = df_, ziformula=~poly(severity_div10, degree=2)+ ExID +scale(WOA) + (1|HenID)+(1|date),
                family="truncated_nbinom2", control=glmmTMBControl(optimizer=optim,optArgs=list(method="BFGS"))) 
summary(ZI_mod_TRB2P)
sim_res = simulateResiduals(ZI_mod_TRB2P)
plot(sim_res)

testDispersion(sim_res)
#pval>0.05 good!

par(mfrow = c(2,2))
plotResiduals(sim_res, df_$WOA)
plotResiduals(sim_res, df_$severity_div10)

#DHARMa also has a special test for zero-inflation, which compares the distribution of expected zeros in the data against the 
#observed zeros
#https://cran.r-project.org/web/packages/DHARMa/vignettes/DHARMa.html#zero-inflation-k-inflation-or-deficits
testZeroInflation(sim_res)
```

```r
#glmer assumption with dharma: https://cran.r-project.org/web/packages/DHARMa/vignettes/DHARMa.html
mod = glmer(WentinWG ~ poly(severity_div10, degree=2) +ExID+scale(WOA) + (1|PenID/HenID)+(1|date), 
           data=df_, family=binomial, control=glmerControl(optimizer="bobyqa", optCtrl=list(maxfun=1e+05)))
summary(mod)
simulationOutput = simulateResiduals(fittedModel = mod, plot = F)
plot(simulationOutput)

effect_plot(mod, pred=severity_div10, interval=TRUE, plot.points=TRUE, partial.residuals=FALSE, data=df_)+xlab("severity")+ylab('went in WG')
#ggsave(file.path(path_,'Severity_Predictability_mean.png'), width=4, height=4, bg = "transparent")
```
