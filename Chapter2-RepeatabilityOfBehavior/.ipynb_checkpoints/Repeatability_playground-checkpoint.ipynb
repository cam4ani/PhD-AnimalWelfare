## uncommon_beh_rep

```r
uncommon_beh_rep = rpt(weekly_has_uncommon_behavior ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) + (1|HenID), 
                       grname = c("HenID"), data = df, datatype = "Binary", link = "logit", adjusted =TRUE,
                       nboot = 1000, npermut = 0)
print(uncommon_beh_rep)
#probit doe snot gave the original-scale, while logit does.
```

```r
#(1|PenID/HenID) +(1|date_HA/date)
#https://rdrr.io/cran/lme4/man/glmer.nb.html:  generalized linear mixed-effects model (GLMM) for the negative binomial family
set.seed(1234)
#faily is not an argument anymore as already nb!
NBMM_GLMER = glmer(weekly_has_uncommon_behavior ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) +(1|HenID), 
                      family=binomial(link = "logit"),data=df)
#, control=glmerControl(optimizer="bobyqa", optCtrl=list(maxfun=1e+05))
summary(NBMM_GLMER)
sim_res = simulateResiduals(NBMM_GLMER, n=1000,plot=T)
```

```r
#repeatabilti yon latent scale:
#https://stats.stackexchange.com/questions/128750/residual-variance-for-glmer
2.935/(2.935+(pi*pi)/3)

#missing value!
#2.935/0.422-2.935-pi*pi/3 #0.730108169621083
2.935/(2.935+(pi*pi)/3+0.730108169621083)
```

```r
ZI_mod_TRB2P = glmmTMB(weekly_nbr_uncommon_behavior ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) +(1|HenID),
                data = df, ziformula=~1,
                family="nbinom1", control=glmmTMBControl(optimizer=optim,optArgs=list(method="BFGS"))) 
summary(ZI_mod_TRB2P)
sim_res = simulateResiduals(ZI_mod_TRB2P)
plot(sim_res)

testDispersion(sim_res)
#pval>0.05 good!

par(mfrow = c(2,2))
plotResiduals(sim_res, df$WOA)
#DHARMa also has a special test for zero-inflation, which compares the distribution of expected zeros in the data against the 
#observed zeros
#https://cran.r-project.org/web/packages/DHARMa/vignettes/DHARMa.html#zero-inflation-k-inflation-or-deficits
testZeroInflation(sim_res)
```

```r
#(1|PenID/HenID) +(1|date_HA/date)
#https://rdrr.io/cran/lme4/man/glmer.nb.html:  generalized linear mixed-effects model (GLMM) for the negative binomial family
set.seed(101)
#faily is not an argument anymore as already nb!
NBMM_GLMER = glmer.nb(weekly_nbr_uncommon_behavior ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) +(1|HenID), 
                      data=df)
#, control=glmerControl(optimizer="bobyqa", optCtrl=list(maxfun=1e+05))
summary(NBMM_GLMER)
sim_res = simulateResiduals(NBMM_GLMER, n=1000,plot=T)

testDispersion(sim_res)
#pval>0.05 good!

par(mfrow = c(2,2))
plotResiduals(sim_res, df$WOA)

#DHARMa also has a special test for zero-inflation, which compares the distribution of expected zeros in the data against the 
#observed zeros
#https://cran.r-project.org/web/packages/DHARMa/vignettes/DHARMa.html#zero-inflation-k-inflation-or-deficits
testZeroInflation(sim_res)
```

```r
distance_travelled = rptPoisson(weekly_avertical_travel_distance ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) + (1|HenID), 
                                grname = c("HenID"), data = df, link = c("log"), adjusted =TRUE, nboot = 1000, npermut = 0)
print(distance_travelled)
#or
#m1 = glmer(weekly_avertical_travel_distance ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) +(1|HenID), 
#          family = poisson, df)
#summary(m1)
#sim_res = simulateResiduals(m1, n=1000,plot=T)
#simulateResiduals(distance_travelled, n=1000,plot=T) : Darma does not support this model!
```

```r
plot(distance_travelled)
```

```r
model_ = lmer(sqrt(weekly_avertical_travel_distance) ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) +(1|HenID), df)
summary(model_)
r.squaredGLMM(model_)
par(mar = c(2, 2, 2, 2))
qqnorm(residuals(model_), main = NULL) 
qqline(residuals(model_))
hist(residuals(model_))
hist(sqrt(df$weekly_avertical_travel_distance))
plot(model_)

# calculate repeatability
set.seed(1)
simulated <- sim(model_, n.sim = 1000)
posterior_HenID <- apply(simulated@ranef$"HenID"[, , 1], 1, var)
posterior_residual <- simulated@sigma^2
quantile(posterior_HenID /
  (posterior_HenID + posterior_residual),
prob = c(0.025, 0.5, 0.975)
)
```
