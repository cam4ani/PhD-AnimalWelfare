# KBF

### vertical travel distance

```r
#https://www.rdocumentation.org/packages/lcsm/versions/0.1.2/topics/fit_bi_lcsm
# Fit bivariate lcsm and save the results 
bi_lavaan_results <- fit_bi_lcsm(data = df_, 
                                 var_y = c("vertical_travel_distance_HA2", "vertical_travel_distance_HA4", "vertical_travel_distance_HA5"),
                                 var_x = c("severity_HA2", "severity_HA4", "severity_HA5"),
                                 model_x = list(alpha_constant = TRUE, 
                                                beta = TRUE,  #not sure!!
                                                phi = FALSE),
                                 model_y = list(alpha_constant = TRUE, 
                                                beta = TRUE, 
                                                phi = TRUE),
                                 coupling = list(delta_lag_xy = TRUE, 
                                                 xi_con_yx = TRUE))

# Save the lavaan syntax that is used to create the layout matrix for semPlot
bi_lavaan_syntax <- fit_bi_lcsm(data = df_, 
                                 var_y = c("vertical_travel_distance_HA2", "vertical_travel_distance_HA4", "vertical_travel_distance_HA5"),
                                 var_x = c("severity_HA2", "severity_HA4", "severity_HA5"),
                                model_x = list(alpha_constant = TRUE, 
                                               beta = TRUE, 
                                               phi = FALSE),
                                model_y = list(alpha_constant = TRUE, 
                                               beta = TRUE, 
                                               phi = TRUE),
                                 coupling = list(delta_lag_xy = TRUE, 
                                                 xi_con_yx = TRUE),
                                return_lavaan_syntax = TRUE,
                                change_letter_x = "M",
                                change_letter_y = "H")

#result
summary(bi_lavaan_results)

# Plot the results
plot_lcsm(lavaan_object = bi_lavaan_results, 
          lavaan_syntax = bi_lavaan_syntax,
          lcsm_colours = TRUE,
          lcsm = "bivariate")
```

```r
#without HA 3: keep 15 weeks in between
#https://www.rdocumentation.org/packages/lcsm/versions/0.1.2/topics/fit_bi_lcsm
# Fit bivariate lcsm and save the results 
#estimator find list here MLR, though we want FILM: Full Information Maximum Likelihood (FIML)  as in the dev. cognitive paper
#to ebst handle missing data, but it does not exist in lavaan,  we cant use it (from my understanding). 
# "The only exception is when data are missing and full information ML is used (via missing = "ML")."
bi_lavaan_results <- fit_bi_lcsm(data = df_, 
                                 var_y = c("WentinWG_HA2", "WentinWG_HA4", "WentinWG_HA5"),
                                 var_x = c("severity_div10_HA2", "severity_div10_HA4", "severity_div10_HA5"),
                                 model_x = list(alpha_constant = TRUE, 
                                                beta = TRUE,  #not sure!!
                                                phi = FALSE),
                                 model_y = list(alpha_constant = TRUE, 
                                                beta = TRUE, 
                                                phi = TRUE),
                                 coupling = list(delta_lag_xy = TRUE, 
                                                 xi_con_yx = TRUE),
                                 missing = "ML")

# Save the lavaan syntax that is used to create the layout matrix for semPlot
bi_lavaan_syntax <- fit_bi_lcsm(data = df_, 
                                var_y = c("WentinWG_HA2", "WentinWG_HA4", "WentinWG_HA5"),
                                 var_x = c("severity_div10_HA2", "severity_div10_HA4", "severity_div10_HA5"),
                                 model_x = list(alpha_constant = TRUE, 
                                                beta = TRUE,  #not sure!!
                                                phi = FALSE),
                                 model_y = list(alpha_constant = TRUE, 
                                                beta = TRUE, 
                                                phi = TRUE),
                                 coupling = list(delta_lag_xy = TRUE, 
                                                 xi_con_yx = TRUE),
                                return_lavaan_syntax = TRUE,
                                change_letter_x = "M",
                                change_letter_y = "H",
                                missing = "ML")

#result
summary(bi_lavaan_results)

# Plot the results
plot_lcsm(lavaan_object = bi_lavaan_results, 
          lavaan_syntax = bi_lavaan_syntax,
          lcsm_colours = TRUE,
          lcsm = "bivariate")
```

```r
#without HA 3: keep 15 weeks in between
#https://www.rdocumentation.org/packages/lcsm/versions/0.1.2/topics/fit_bi_lcsm
# Fit bivariate lcsm and save the results 
bi_lavaan_results <- fit_bi_lcsm(data = df, 
                                 var_y = c("WentinWG_HA2", "WentinWG_HA4", "WentinWG_HA5"),
                                 var_x = c("FD_HA2", "FD_HA4", "FD_HA5"),
                                 model_x = list(alpha_constant = TRUE, 
                                                beta = TRUE,  #not sure!!
                                                phi = FALSE),
                                 model_y = list(alpha_constant = TRUE, 
                                                beta = TRUE, 
                                                phi = TRUE),
                                 coupling = list(xi_con_yx = TRUE, 
                                                 xi_con_xy = TRUE),
                                estimator='mlr',missing='fiml')

# Save the lavaan syntax that is used to create the layout matrix for semPlot
bi_lavaan_syntax <- fit_bi_lcsm(data = df, 
                                var_y = c("WentinWG_HA2", "WentinWG_HA4", "WentinWG_HA5"),
                                var_x = c("FD_HA2", "FD_HA4", "FD_HA5"),
                                model_x = list(alpha_constant = TRUE, 
                                               beta = TRUE, 
                                               phi = FALSE),
                                model_y = list(alpha_constant = TRUE, 
                                               beta = TRUE, 
                                               phi = TRUE),
                                coupling = list(xi_con_yx = TRUE,
                                                xi_con_xy = TRUE), #estimate whether changes in x at time t are predicting subsequent changes in y at time t+1
                                return_lavaan_syntax = TRUE,
                                change_letter_x = "M",
                                change_letter_y = "H",
                               estimator='mlr',missing='fiml')

#result
summary(bi_lavaan_results)

# Plot the results
plot_lcsm(lavaan_object = bi_lavaan_results, 
          lavaan_syntax = bi_lavaan_syntax,
          lcsm_colours = TRUE,
          lcsm = "bivariate")
```
