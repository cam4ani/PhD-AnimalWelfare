```r
library(rptR)
library(ggplot2)
library(lindia)
library(tidyverse)
library(lmerTest) #lmer, glmer
library(DHARMa) #verify assumption of glm
library(merTools) #sim
library(MuMIn) #r2
library(glmmTMB)
library(glmmADMB)
library(lme4)
library(MASS)
library(boot) #bootstrap
```

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/ALLDATA_'
df = read.csv(file.path(path_,'df_BS.csv'), header = TRUE, sep = ",")
df$HenID = as.factor(df$HenID)   
df$PenID = as.factor(df$PenID)
df$TrackingSystemID = as.factor(df$TrackingSystemID)
df$level = as.factor(df$level) 
df$FoodPenID = as.factor(df$FoodPenID) 
df$CLASS = as.factor(df$CLASS) 
df$Treatment = as.factor(df$Treatment)

#setting reference group
contrasts(df$Treatment)
df <- df %>% mutate(Treatment = relevel(Treatment, ref = "TRAN"))
contrasts(df$Treatment)

contrasts(df$CLASS)
df <- df %>% mutate(CLASS = relevel(CLASS, ref = "LEXP"))
contrasts(df$CLASS)
print(dim(df))
summary(df)

df$latency_1_Zone_mn_6hOPP = max(df$latency_1_Zone_mn_6h)-df$latency_1_Zone_mn_6h
df$food_related_behavior_final_inv_cst = (-1)*(df$food_related_behavior_final)+1.01
df$latency_1_Zone_15mn_6h = as.integer(df$latency_1_Zone_mn_6h/15)
df$latency_1_Zone_15mn_6hOPP = max(df$latency_1_Zone_15mn_6h) - df$latency_1_Zone_15mn_6h
df$latency_1_Zone_15mn_3h = as.integer(df$latency_1_Zone_mn_3h/15)
df$latency_1_Zone_15mn_3hOPP = max(df$latency_1_Zone_15mn_3h) - df$latency_1_Zone_15mn_3h
df$perc_1_Zone_while_WG_open = as.integer(df$perc_1_Zone_while_WG_open)

df_time = df[df$BS_type=='time',]
dim(df_time)

df_LP2 = df[df$BS_type=='LP2',]
dim(df_LP2)

df_LP10 = df[df$BS_type=='LP10',]
dim(df_LP10)

df_cold = df[df$BS_type=='cold',]
dim(df_cold)

df_vacc = df[df$BS_type=='TypicalVaccinationDay',]
dim(df_vacc)

df_AC = df[df$across_context==1,]
dim(df_AC)
```

# 1. Sleeping_NotUppest

#### PenID

```r
glmer(Sleeping_NotUppest ~  scale(weight_interp)+scale(KBF_interp) + (1|PenID/HenID), data=df_time, family='binomial',
      control = glmerControl(optimizer ='optimx', optCtrl=list(method='nlminb')))
#penid as as random effect --> no cvg, penID have 0 var
```

```r
#PenID as fixed effect
mod__1 = glmer(Sleeping_NotUppest ~  scale(weight_interp)+scale(KBF_interp) +(1|HenID), data=df_time, family='binomial',
      control = glmerControl(optimizer ='optimx', optCtrl=list(method='nlminb')))

mod__2 = glmer(Sleeping_NotUppest ~  scale(weight_interp)+scale(KBF_interp) + PenID+(1|HenID), data=df_time, family='binomial',
      control = glmerControl(optimizer ='optimx', optCtrl=list(method='nlminb')))
#cant do anova for binomial
#--> cant use, no var
summary(mod__1)
summary(mod__2)
AIC(mod__1,mod__2)
r.squaredGLMM(mod__1)
r.squaredGLMM(mod__2)
```

```r
#by curiosity lets check the repeatabitlily of penid having also the henid
set.seed(1234)
mod_ = rpt(Sleeping_NotUppest ~ Treatment + scale(weight_interp) + scale(KBF_interp) + (1|PenID)+(1|HenID),
               grname=c('PenID',"HenID"), datatype="Binary", data=df_time, adjusted=TRUE, nboot=5, npermut=0,
               link='logit') 
mod_
sim_res = simulateResiduals(mod_$mod, n=1000)
plot(sim_res)
plot(mod_, grname = "HenID", type = "boot", cex.main = 0.8)
#--> R=0, singular value
```

```r
############ GAMMA
modF = glmmTMB(mid_cum_Z4_h_Morning ~ Treatment + scale(weight_interp) + scale(KBF_interp) + scale(DIB) + (1|HenID), 
                family = Gamma(link="log"), data=df_time)
summary(modF)
sim_res = simulateResiduals(modF, n=500)
plot(sim_res)
testOutliers(sim_res)
#plot residual against predictors
plotResiduals(sim_res, form = df_time$DIB)
plotResiduals(sim_res, form = df_time$KBF_interp)
plotResiduals(sim_res, form = df_time$weight_interp)
plotResiduals(sim_res, form = df_time$Treatment)
#fit perfectly!
```

```r
############ GAUSSIAN (with transformation sqrt())
mod_ = rpt(sqrt(mid_cum_Z4_h_Morning) ~ Treatment + scale(weight_interp) + scale(KBF_interp) + scale(DIB) +(1|HenID), 
                       grname = c("HenID"), data = df_time, datatype = "Gaussian", adjusted =TRUE,
                       nboot = nboot_, npermut = 0)
print(mod_)
model_ = mod_$mod
summary(model_)
par(mar = c(2, 2, 2, 2))
qqnorm(residuals(model_), main = NULL) 
qqline(residuals(model_))
hist(residuals(model_))
plot(model_)
sim_res = simulateResiduals(model_, n=1000)
plot(sim_res)
```

```r
############ CHECK PENID
#with gaussian
mod_1 = lmer(scale(sqrt(mid_cum_Z4_h_Morning)) ~ Treatment + scale(weight_interp) + scale(KBF_interp) + scale(DIB) +(1|HenID), REML=TRUE, 
            data = df_time)
summary(mod_1)
#0.01814   /(0.01814    + 0.43923   + 0.46052  )*100 #=1.97627166653956
mod_2 = lmer(scale(sqrt(mid_cum_Z4_h_Morning)) ~ Treatment + scale(weight_interp) + scale(KBF_interp) + scale(DIB) +(1|PenID/HenID), REML=TRUE, 
            data = df_time)
summary(mod_2)
AIC(mod_1, mod_2)

#with gamma
modF1 = glmmTMB(mid_cum_Z4_h_Morning ~ Treatment + scale(weight_interp) + scale(KBF_interp) + scale(DIB) + (1|HenID), 
                family = Gamma(link="log"), data=df_time)
summary(modF1)
#without accounting for residual: 0.0004062/(0.0208171 +0.0004062)*100 = 1.91393421381218
modF2 = glmmTMB(mid_cum_Z4_h_Morning ~ Treatment + scale(weight_interp) + scale(KBF_interp) + scale(DIB) + (1|PenID/HenID), 
                family = Gamma(link="log"), data=df_time)
summary(modF2)
AIC(modF1, modF2)
#--> dont put PenID
```
