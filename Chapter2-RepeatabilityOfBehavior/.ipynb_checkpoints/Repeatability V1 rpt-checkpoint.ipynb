## uncommon_beh_rep

```r
has_uncommon_beh = rpt(weekly_has_uncommon_behavior ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) + (1|HenID), 
                       grname = c("HenID"), data = df, datatype = "Binary", link = "logit", adjusted =TRUE,
                       nboot = nboot_, npermut = npermut_)
print(has_uncommon_beh)
#probit does not gave the original-scale, while logit does.

#penid singular: grname = c("HenID:PenID") #as in glmer
#same results as with:
#mod_ = glmer(weekly_has_uncommon_behavior ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) +(1|HenID), 
#                      family=binomial(link = "logit"),data=df)
#but not latent scale repeatabitliy:#https://stats.stackexchange.com/questions/128750/residual-variance-for-glmer 
#2.935/(2.935+(pi*pi)/3) = 0.471495931634641
#no assumption: https://www.statisticssolutions.com/free-resources/directory-of-statistical-analyses/assumptions-of-logistic-regression/
sim_res = simulateResiduals(has_uncommon_beh$mod, n=1000)
plot(sim_res)
testOutliers(sim_res, type = 'bootstrap')
plotResiduals(sim_res, df$WOA, quantreg = T)
#plot shows the simulated number of zeroes with your actual number shown as red. You have fewer zeroes than expected if the red
#line is on the left
testZeroInflation(sim_res)
#This function performs a simulation-based test for over/underdispersion
#plot shows your actual dispersion with a histogram of simulated residuals. 
#Since the red  value is to the left it indicates that the model has under-dispersion
testDispersion(sim_res)
```

### proportion of day sleeping up

```r
#as in beetle examples
set.seed(1234)
mod_ = rpt(cbind(weekly_nbr_Is_Sleeping_UP, weekly_nbr_Is_Sleeping_NOTUP) ~ Treatment + scale(weekly_avg_temperature) + scale(WOA) +scale(I(WOA^2)) + (1|HenID),
               grname=c("HenID"), datatype = "Proportion", data=df, nboot = nboot_, npermut = npermut_) 
mod_
sim_res = simulateResiduals(mod_$mod, n=1000)
plot(sim_res)
#tests if the overall distribution conforms to expectations
testUniformity(sim_res)
#fits a quantile regression or residuals against a predictor (default predicted value), and tests of this conforms to the 
#expected quantile
testQuantiles(sim_res)

#against predictors
plotResiduals(sim_res, df$WOA, quantreg = T)
plotResiduals(sim_res, df$weekly_avg_temperature, quantreg = T)
testCategorical(sim_res, catPred = df$Treatment) #more info than: plotResiduals(sim_res, df$Treatment, quantreg = T)

#tests if there are more simulation outliers than expected
testOutliers(sim_res)

#plot shows the simulated number of zeroes with your actual number shown as red. You have fewer zeroes than expected if the red
#line is on the left
testZeroInflation(sim_res)

#tests if the simulated dispersion is equal to the observed dispersion
#This function performs a simulation-based test for over/underdispersion
#plot shows your actual dispersion with a histogram of simulated residuals. 
#Since the red  value is to the left it indicates that the model has under-dispersion
testDispersion(sim_res)
```

```r
#or gaussian
distance_travelled = rpt(weekly_avertical_travel_distance ~ Treatment + scale(weekly_avg_temperature) + scale(WOA)+ scale(I(WOA^2)) + (1|HenID), 
                       grname = c("HenID"), data = df, datatype = "Gaussian", adjusted =TRUE,
                       nboot = nboot_, npermut = npermut_)
print(distance_travelled)
model_ = distance_travelled$mod
par(mar = c(2, 2, 2, 2))
qqnorm(residuals(model_), main = NULL) 
qqline(residuals(model_))
hist(residuals(model_))
plot(model_)
```

```r
#or best but latent scale:
model_ = lmer(sqrt(weekly_avertical_travel_distance) ~ Treatment + scale(weekly_avg_temperature) + scale(WOA)+ scale(I(WOA^2)) +(1|HenID), df)
summary(model_)
r.squaredGLMM(model_)
par(mar = c(2, 2, 2, 2))
qqnorm(residuals(model_), main = NULL) 
qqline(residuals(model_))
hist(residuals(model_))
hist(sqrt(df$weekly_avertical_travel_distance))
plot(model_)

# calculate repeatability
set.seed(1)
simulated <- sim(model_, n.sim = 1000)
posterior_HenID <- apply(simulated@ranef$"HenID"[, , 1], 1, var)
posterior_residual <- simulated@sigma^2
quantile(posterior_HenID /
  (posterior_HenID + posterior_residual),
prob = c(0.025, 0.5, 0.975)
)
```
