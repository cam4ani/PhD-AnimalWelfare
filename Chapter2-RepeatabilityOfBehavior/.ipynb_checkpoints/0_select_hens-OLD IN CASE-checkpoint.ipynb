```python
#compute repeated measure correlation
for i in range(0,len(li_mvt_4visual)):
    beh1 = li_mvt_4visual[i]
    for beh2 in li_mvt_4visual[i+1:]:
        print('----------------------------------------------------------------------------')
        print(beh1, '   ', beh2)
        print('----------------------------------------------------------------------------')
        #due to non-normality of all behaviors except vertical travel distance, we will use the spearman rank and kendalltau
        #correaltions, but due to repeated measurement, we will do it for each HAID
        for haID, df_haid in df_lcsm_.groupby(['date_HA']):
            print(' ')
            print(haID, df_haid.shape)
            print(pearsonr(df_haid[beh1].tolist(), df_haid[beh2].tolist()))
            print(spearmanr(df_haid[beh1].tolist(), df_haid[beh2].tolist()))
            print(kendalltau(df_haid[beh1].tolist(), df_haid[beh2].tolist()))
        try:
            print(pg.rm_corr(data=df_lcsm_[df_lcsm_['HenID'].isin(li_hen2keep)], x=beh1, y=beh2, subject='HenID'))
            print(pg.plot_rm_corr(data=df_lcsm_[df_lcsm_['HenID'].isin(li_hen2keep)], x=beh1, y=beh2, subject='HenID'));
            plt.savefig(os.path.join(path_extracted_data_chapter2,'ReapCorr_'+beh1+'_'+beh2+'.png'), bbox_inches='tight')
            plt.show();
        except Exception as e:
            print(e)
```
