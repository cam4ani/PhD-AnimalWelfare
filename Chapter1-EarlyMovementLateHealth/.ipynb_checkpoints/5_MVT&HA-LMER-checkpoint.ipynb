```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/ALLDATA_/Adaptability'
#df = read.csv(file.path(path_,'Adaptability_Stat.csv'), header = TRUE, sep = ";") #old
df = read.csv(file.path(path_,'df_MVT_4stat.csv'), header = TRUE, sep = ";")
#df$'#of the first 3 days with 4 zone visited' = df$nbr_days_4zonevisited
#df$'mean % of time spent in lower perch' = df$mean_perc_duration_3_Zone
#df$'mean % of time spent in WG' = df$mean_percentage_time_outside_while_WG_open
#df$Person_feather = as.factor(df$Person_feather)   
df$PenID = as.factor(df$PenID) 
df$RPen = as.factor(df$RPen) 
df$Treatment = as.factor(df$Treatment) 
df$CLASS = as.factor(df$CLASS) 
df$perc_1_Zone_while_WG_open_10 = df$perc_1_Zone_while_WG_open/10
df$perc_duration_5_Zone_10 = df$perc_duration_5_Zone /10
print(dim(df))
summary(df)
head(df,3)
```

# Linear mixed effect model

```r
#FEATHERS
#test other fixed effects
#we will keep the treatment, and believe that it will have all info needed from rearing pen, laying pen. 
#CLASS might not be reliable
modeltestinit = lmer(Feathers ~ 1 + (1|PenID), REML = FALSE, data = df)
summary(modeltestinit)
r.squaredGLMM(modeltestinit)

modeltest1 = lmer(Feathers ~ Treatment + (1|PenID), REML = FALSE, data = df)
summary(modeltest1)
r.squaredGLMM(modeltest1)

modeltest2 = lmer(Feathers ~  CLASS + (1|PenID), REML = FALSE, data = df)
summary(modeltest2)
r.squaredGLMM(modeltest2)

modeltest3 = lmer(Feathers ~ RPen + (1|PenID), REML = FALSE, data = df)
summary(modeltest3)
r.squaredGLMM(modeltest3)

modeltest3 = lmer(Feathers ~ InitialWeight_kg + (1|PenID), REML = FALSE, data = df)
summary(modeltest3)
r.squaredGLMM(modeltest3)

summary(modeltestinit)$logLik
summary(modeltest1)$logLik
summary(modeltest2)$logLik
summary(modeltest3)$logLik

#The lower the AIC value, the more parsimonious the fit of the model is to the data
anova(modeltest1, modeltestinit, test="Chisq")
anova(modeltest2, modeltestinit, test="Chisq")
anova(modeltest3, modeltestinit, test="Chisq")
#LRT : p>0.05--> use simpler model
```

```r
#KBF SEVERITY
#test other fixed effects
#we will keep the treatment, and believe that it will have all info needed from rearing pen, laying pen. 
#CLASS might not be reliable
modeltestinit = lmer(severity ~ 1 + (1|PenID), REML = FALSE, data = df)
summary(modeltestinit)
r.squaredGLMM(modeltestinit)

modeltest1 = lmer(severity ~ Treatment + (1|PenID), REML = FALSE, data = df)
summary(modeltest1)
r.squaredGLMM(modeltest1)

modeltest2 = lmer(severity ~  CLASS + (1|PenID), REML = FALSE, data = df)
summary(modeltest2)
r.squaredGLMM(modeltest2)

modeltest3 = lmer(severity ~ RPen + (1|PenID), REML = FALSE, data = df)
summary(modeltest3)
r.squaredGLMM(modeltest3)

modeltest3 = lmer(severity ~ InitialWeight_kg + (1|PenID), REML = FALSE, data = df)
summary(modeltest3)
r.squaredGLMM(modeltest3)

summary(modeltestinit)$logLik
summary(modeltest1)$logLik
summary(modeltest2)$logLik
summary(modeltest3)$logLik

#The lower the AIC value, the more parsimonious the fit of the model is to the data
anova(modeltest1, modeltestinit, test="Chisq")
anova(modeltest2, modeltestinit, test="Chisq")
anova(modeltest3, modeltestinit, test="Chisq")
#LRT : p>0.05--> use simpler model
```

# Feature selection via LASSO

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/ALLDATA_/Adaptability'
df_FS = read.csv(file.path(path_,'df_MVT_4Featureselect.csv'), header = TRUE, sep = ";")
print(dim(df_FS))
summary(df_FS)
head(df_FS,3)
```

```r
set.seed(123)    
li_varF = c('nbr_first3days_no_mvt','perc_duration_5_Zone','perc_1_Zone_while_WG_open','vertical_travel_distance_perh',
            'BLUP_slopes1','BLUP_slopes2','Predictability_mean','BehaviouralType')

df_FS_ = df_FS[c(li_varF,c('Feathers'))]
df_FS_ = na.omit(df_FS_)
dim(df_FS_)
y = df_FS_ %>% select(Feathers) %>% as.matrix() #'Feathers','severity','weight'
X = df_FS_ %>% select(li_varF) %>% as.matrix()

cvfit = glmnet(X, y, type.measure = "mse", alpha=1)
#nonzero coefficients (Df), the percent (of null) deviance explained (%dev) and the value of λ (Lambda). A
print(cvfit)
plot(cvfit, label=TRUE)
#coef(cvfit, s = "lambda.min")
#coef(cvfit, s = "lambda.1se")
coef(cvfit)
#--> perc_1_Zone_while_WG_open,  Predictability_mean, vertical_travel_distance_perh, nbr_first3days_no_mvt
```

```r
set.seed(123)    
li_varF = c('nbr_first3days_no_mvt','perc_duration_5_Zone','perc_1_Zone_while_WG_open','vertical_travel_distance_perh',
            'BLUP_slopes1','BLUP_slopes2','Predictability_mean','BehaviouralType')

df_FS_ = df_FS[c(li_varF,c('severity'))]
df_FS_ = na.omit(df_FS_)
dim(df_FS_)
y = df_FS_ %>% select(severity) %>% as.matrix() #'Feathers','severity','weight'
X = df_FS_ %>% select(li_varF) %>% as.matrix()

cvfit = glmnet(X, y, type.measure = "mse", alpha=1)
#nonzero coefficients (Df), the percent (of null) deviance explained (%dev) and the value of λ (Lambda). A
print(cvfit)
plot(cvfit, label=TRUE)
#coef(cvfit, s = "lambda.min")
#coef(cvfit, s = "lambda.1se")
coef(cvfit)
#--> nbr_first3days_no_mvt, BLUP_slopes2, perc_duration_5_Zone, vertical_travel_distance_perh
```

```r
set.seed(123)    
li_varF = c('nbr_first3days_no_mvt','perc_duration_5_Zone','perc_1_Zone_while_WG_open','vertical_travel_distance_perh',
            'BLUP_slopes1','BLUP_slopes2','Predictability_mean','BehaviouralType')

df_FS_ = df_FS[c(li_varF,c('weight'))]
df_FS_ = na.omit(df_FS_)
dim(df_FS_)
y = df_FS_ %>% select(weight) %>% as.matrix() #'Feathers','severity','weight'
X = df_FS_ %>% select(li_varF) %>% as.matrix()

cvfit = glmnet(X, y, type.measure = "mse", alpha=1)
#nonzero coefficients (Df), the percent (of null) deviance explained (%dev) and the value of λ (Lambda). A
print(cvfit)
plot(cvfit, label=TRUE)
#coef(cvfit, s = "lambda.min")
#coef(cvfit, s = "lambda.1se")
coef(cvfit)
#--> BLUP_slopes1, nbr_first3days_no_mvt, Predictability_mean, perc_duration_5_Zone
```

# Feathers

```r
#--> perc_1_Zone_while_WG_open,  Predictability_mean, vertical_travel_distance_perh 
```

```r
hist(df$Feathers)
```

```r
#only two groups so treatment as fixed effect (otherwise (1|Treatment/PenID) lead to boundary (singular) fit: see ?isSingular)
#for predictive plot in poster
fit = lmerTest::lmer(Feathers ~ perc_1_Zone_while_WG_open_10 + vertical_travel_distance_perh + Predictability_mean +nbr_first3days_no_mvt + (1|PenID), 
                     data=df)
fit_feather = fit
anova(fit)
effCoef(fit)
r.squaredGLMM(fit)

#capture.output(summary(fit), file=file.path(path_,paste(name,'.txt', sep='')), append=TRUE)
summary(fit) #, default: ddf = "Satterthwaite", other: "Kenward-Roger"
#Assumptions
###### normally distributed residuals
#qqmath(fit, id=0.05)
qqnorm(resid(fit), main = "Residuals")
qqline(resid(fit))
###### normally of random effects
qqmath(ranef(fit,postVar = TRUE))
######check homogeneity of variance (residuals has constant variance)
#variance of the residuals is equal across groups. Even spread around the centered line?
plot(fit)

#plot estimates
plot_model(fit, vline.color = "red")
```

```r
#profile likelihood confidence intervals
confint(fit_feather)
```

```r
tab_model(fit_feather, file=file.path(path_,'LMERFeather.html'))
# then take this html file and make .png file
webshot(file.path(path_,'LMERFeather.html'), file.path(path_,'LMERFeather.png'), zoom=3, cliprect = c(0, 5, 600, 270))
```

```r
sjPlot::plot_model(fit_feather, type = "pred", terms = c("perc_1_Zone_while_WG_open_10"), title='')+ xlab("Mean percentage of time spent in winter garden")+ylab('Feather coverage')+
      theme(plot.background = element_rect(fill = "transparent",colour = NA),
           axis.text.x = element_text(color="black"), axis.ticks = element_line(color = "black"),
           axis.text.y = element_text(color="black"))
ggsave(file.path(path_,'Feather_WG_pred.png'), width=4, height=4, bg = "transparent")
```

```r
ggscatter(df, x = "perc_1_Zone_while_WG_open_10", y = "Feathers",color = "PenID")
ggsave(file.path(path_,'Feather_WG_scatter.png'),width=5, height=5)
```
