# Models

# Behavioural type

```r
modelBT = lmer(PC1 ~ cDIB + RearingPenID + Treatment + CLASS + InitialWeight_kg + temperature_C_avg + (1|PenID/HenID), 
               REML = FALSE, data = df)
summary(modelBT)
r.squaredGLMM(modelBT)
```

```r
###### normally distributed residuals
qqnorm(resid(modelBT))
qqline(resid(modelBT))
hist(resid(modelBT))
######check homogeneity of variance (residuals has constant variance)
plot(modelBT)
```

```r
#repeatabilities
print(VarCorr(modelBT),comp =c("Variance","Std.Dev."))
rep = VarCorr(modelBT)$"HenID:PenID"[1]/
(VarCorr(modelBT)$"HenID:PenID"[1]+
 VarCorr(modelBT)$"PenID"[1]+
 attr(VarCorr(modelBT), "sc")^2) # equal to the varance of residual
#alternatively: 0.1027/(0.1027+0.000086313+0.11594)
rep
```

```r
#consistent among pen variation, i.e. if birds PC1 is consistently more or less in some pens
#birds do not show consisten moement variation between pens(PenID explains ~0.04% of the variation in the mvt)
#we may therefore simplify our mode lby removing the random intercept for PenID
VarCorr(modelBT)$"PenID"[1]/
(VarCorr(modelBT)$"HenID:PenID"[1]+
 VarCorr(modelBT)$"PenID"[1]+
 attr(VarCorr(modelBT), "sc")^2) 
```

```r
set.seed(1)
simulated = sim(modelBT, n.sim = 1000)
posterior_HenID = apply(simulated@ranef$"HenID:PenID"[ , , 1],1,var)
posterior_PenID = apply(simulated@ranef$"PenID"[ , , 1],1,var)
posterior_residual  = simulated@sigma^2
#credible interval around our repeatability estimate for individual variation
#on average 49% of the remianing variance (after controlling or ext. temperature, weight and rearing pen identity) 
#in PC1 can be attributed to differences between individuals, suggesting that some birds always have lower PC1 
#compared to other birds, and this differenc eis naot cause by preditable weight, temeprature or rearign pen variation.
quantile(posterior_HenID/
         (posterior_HenID+posterior_PenID+posterior_residual),
         prob=c(0.025, 0.5, 0.975))
quantile(posterior_PenID/
         (posterior_HenID+posterior_PenID+posterior_residual),
         prob=c(0.025, 0.5, 0.975))
```

```r
#behavioural type using the "merTools-package to obtain repeated samples from the posterior distribution
#of our model from which we can calculate the standard deviation for each level of the random effect"
randomSims = REsim(modelBT, n.sims = 1000)
head(randomSims[randomSims$groupFctr=="HenID:TrackingSystemID",])
#Use it for visual of behavioural type
```

```r
write.csv(randomSims, file.path(path_,'Adaptability','BehaviouralType.csv'), row.names = FALSE)
```

# Plasticity

```r
#do birds differ in how they shift their general mvements (PC1) across the #days post-transfer to a laying barn?
#individual variation in plasticity
#poly(cDIB, 2, raw = T)
```

```r
#only a random intercept for hens
modelPL1 = lmer(PC1 ~ cDIB + RearingPenID + Treatment + CLASS + InitialWeight_kg + temperature_C_avg + (1|HenID), 
                REML = FALSE, data = df,
                control = lmerControl(optimizer ='optimx', optCtrl=list(method='nlminb')))
summary(modelPL1)
r.squaredGLMM(modelPL1)
```

```r
###### normally distributed residuals
qqnorm(resid(modelPL1))
qqline(resid(modelPL1))
hist(resid(modelPL1))
######check homogeneity of variance (residuals has constant variance)
plot(modelPL1)
```

```r
#random interecepts and slopes over DIB for each animal (interactio between DIB and henID as a random effect)
modelPL2 = lmer(PC1 ~ cDIB + RearingPenID + Treatment + CLASS + InitialWeight_kg + temperature_C_avg + (1 + cDIB|HenID), 
                REML = FALSE, data = df,
                control = lmerControl(optimizer ='optimx', optCtrl=list(method='nlminb')))
summary(modelPL2)
r.squaredGLMM(modelPL2)
```

```r
###### normally distributed residuals
qqnorm(resid(modelPL2))
qqline(resid(modelPL2))
hist(resid(modelPL2))
######check homogeneity of variance (residuals has constant variance)
plot(modelPL2)
```

```r
#assumptions not met!
#test whether a nonlinear random slope over DIB improve model fit
modelPL3 = lmer(PC1 ~ poly(cDIB, 2) + RearingPenID+Treatment+CLASS+InitialWeight_kg+temperature_C_avg +(1+poly(cDIB, 2)|HenID), 
                REML = FALSE, data = df,
                control = lmerControl(optimizer ='optimx', optCtrl=list(method='nlminb')))
summary(modelPL3)
r.squaredGLMM(modelPL3)
```

```r
###### normally distributed residuals
qqnorm(resid(modelPL3))
qqline(resid(modelPL3))
hist(resid(modelPL3))
######check homogeneity of variance (residuals has constant variance)
plot(modelPL3)
```

```r
#compare the log-likelihoods of the two models and perform a likelihood ratio test (LRT).
#Here, we take 2 ×the log-likelihood values from the model summaries to determine a 𝜒+value, 
#and then determine the p-value of the LRT from the 𝜒+value and the difference of the degrees of freedom(df) between the two models using the function 1-pchisq(chi2, df)
#CHECK df difference between models (by looking at the df within the models being compared)
summary(modelPL1)$logLik
summary(modelBT)$logLik
chi2 = 2*(summary(modelBT)$logLik -summary(modelPL1)$logLik)
1-pchisq(chi2,1) # chi2, df
```

```r
#compare the log-likelihoods of the two models and perform a likelihood ratio test (LRT).
#Here, we take 2 ×the log-likelihood values from the model summaries to determine a 𝜒+value, 
#and then determine the p-value of the LRT from the 𝜒+value and the difference of the degrees of freedom(df) between the two models using the function 1-pchisq(chi2, df)
#CHECK df difference between models (by looking at the df within the models being compared)
summary(modelPL1)$logLik
summary(modelPL2)$logLik
chi2 = 2*(summary(modelPL2)$logLik -summary(modelPL1)$logLik)
1-pchisq(chi2,2) # chi2, df
```
