```r
#install.packages("dhglm")
library(dhglm)
library(hglm)
#hglm: to fit hierarchical generalized linear models: those models "extend generalized linear models by relaxing the assumption 
#that error components are independent. This allows models to be built in situations where more than one error term is 
#necessary and also allows for dependencies between error terms (wikipedia)""
```

# Download data

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/ALLDATA_'
path_adapt = file.path(path_,'Adaptability')
path_model = file.path(path_adapt,'repeatability_model')
df = read.csv(file.path(path_adapt,'df_MVT_4individuality.csv'), header = TRUE, sep = ";")
df$HenID = as.factor(df$HenID)   
df$PenID = as.factor(df$PenID) 
df$WIB = as.factor(df$WIB) 
df$TrackingSystemID = as.factor(df$TrackingSystemID) 
#ean-centring of the environmental variable so that intercepts reflect average values for the population and individuals
#df$cDIB = scale(df$temperatute) #TODO
df$cDIB = as.integer(df$DIB) 
df$cDIBfact = as.factor(df$cDIB) 
print(dim(df))
summary(df)
```

```r
head(df)
```

# Estimates predictibility

```r
mod = hglm(fixed = PC1_standardized ~ DIB, random = ~1 | HenID, data = df, family=gaussian(link = identity)) #family = binomial(link = logit), 
summary(mod)
print(summary(mod), print.ranef = TRUE)
#"The variance of the random individual effect was estimated to 0.3764143 and the algorithm converged in 2 iterations."
```

```r
#### Model checking plot for crack-growth data
#PC1_standardized ~ WIB + (1|HenID), data = df
model_mu = DHGLMMODELING(Model="mean", Link="identity", LinPred=PC1_standardized~WIB+(1|HenID),RandDist="gaussian")
model_phi = DHGLMMODELING(Model="dispersion", Link="identity", LinPred=phi~(1|HenID),RandDist="gaussian")
res_crack_dhglm1 = dhglmfit(RespDist="gaussian", DataMain=df, MeanModel=model_mu, DispersionModel=model_phi,Maxiter=1)
plotdhglm(res_crack_dhglm1)
```

```r
summary(res_crack_dhglm1)
```

```r
#### Model checking plot for crack-growth data
data(crack_growth)
model_mu = DHGLMMODELING(Model="mean", Link="log", LinPred=y~crack0+(1|specimen), RandDist="inverse-gamma")
model_phi = DHGLMMODELING(Model="dispersion", Link="log", LinPred=phi~cycle+(1|specimen), RandDist="gaussian")
res_crack_dhglm1 = dhglmfit(RespDist="gamma", DataMain=crack_growth, MeanModel=model_mu, DispersionModel=model_phi,Maxiter=1)
plotdhglm(res_crack_dhglm1)
```

```r
mod = hglm2(meanmodel = PC1_standardized ~ WIB + (1|HenID), data = df, family=gaussian(link = identity)) #family = binomial(link = logit), 
summary(mod)
print(summary(mod), print.ranef = TRUE)
#"The variance of the random individual effect was estimated to 0.3764143 and the algorithm converged in 2 iterations."
```
