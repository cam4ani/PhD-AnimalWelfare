# principal()

```r
#https://www.rdocumentation.org/packages/psych/versions/2.1.9/topics/principal
#df[li_pca] = scale(df[li_pca], center=TRUE, scale=TRUE) #x-mean / sd
res = psych::principal(df_pca, rotate="none", nfactors=4, scores=TRUE, cor='cor') #not varimax (we want only one PC)
res

#documentation: The beta weights to find the principal components from the data
res$weights

#eigenvalues of all components --> first one >1
#from https://stats.oarc.ucla.edu/spss/seminars/introduction-to-factor-analysis/a-practical-introduction-to-factor-analysis/:
#represent the total amount of variance that can be explained by a given principal component (if negative: illconditioned model)
#=sum(x^2), x= standardize laodings: i.e Eigenvalues are also the sum of squared component loadings across all items for each 
#component, which represent the amount of variance in each item that can be explained by the principal component
#Eigenvalues close to zero imply there is item multicollinearity, since all the variance can be taken up by the first component.
res$values
```

```r
sqrt(1/length(li_pca))
```

```r
#SVD (singular value decomposition) used here (vs. Eigen decomposition)
#For computational reasons, PCA is more commonly performed via the singular value decomposition (SVD) 
res = prcomp(df_pca, center = TRUE, scale. = TRUE, retx=FALSE)
#scale. = TRUE: analysis to be performed on a correlation matrix (p.4: http://strata.uga.edu/software/pdf/pcaTutorial.pdf)
res
summary(res)
#R doc: 
#sdev: the standard deviations of the principal components (i.e., the square roots of the eigenvalues of the covariance/
#correlation matrix, though the calculation is actually done with the singular values of the data matrix).
#rotation: the matrix of variable loadings (i.e., a matrix whose columns contain the eigenvectors). The function princomp returns this in the element loadings.
```

# By hand

```r
df_pcastd = scale(df_pca[li_pca], center=TRUE, scale=TRUE) #x-mean / sd
```

```r
#find correlation matrix: correlaiton based PCA
CorMatrix = cor(df_pcastd)
```

```r
#find the eigenvalues and eigenvectors of this correlation matrix through spectral decomposition
eigenObject <- eigen(CorMatrix)
pcaVariances <- eigenObject$values 
pcaLoadings <- eigenObject$vectors 
```

```r
#add col & row names to the eigenvectors (loadings)
rownames(pcaLoadings) <- colnames(df_pcastd) 
PClabels <- paste('PC', 1:ncol(pcaLoadings), sep='') 
colnames(pcaLoadings) <- PClabels
pcaLoadings
```

```r
#daily hens scores
df_std = scale(df[li_pca], center=TRUE, scale=TRUE) #x-mean / sd
df_std_PC = df_std %*% eigenObject$vectors 
colnames(df_std_PC) <- PClabels
df_std_PC
```

```r
#daily hens scores on all dataset
df_std = scale(df[li_pca], center=TRUE, scale=TRUE) #x-mean / sd
df_std_PC = df_std %*% eigenObject$vectors 
colnames(df_std_PC) <- PClabels
df_std_PC
df_std_PC = data.frame(df_std_PC)
df$PC1 = df_std_PC$PC1
df
```
