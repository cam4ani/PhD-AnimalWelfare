#unbalanced dataset: usage of stratified sample
#strata: A (factor) variable that is used for stratified sampling.
#sampsize: Size(s) of sample to draw. For classification, if sampsize is a vector of the length the number of strata, 
#then sampling is stratified by strata, and the elements of sampsize indicate the numbers to be drawn from the strata.
#The overall complexity of RF is something like ntree⋅mtry⋅(# objects)log(# objects) 
#for efficiency: https://stats.stackexchange.com/questions/37370/random-forest-computing-time-in-r
#mtry tuning
#Select mtry variables at random from the p regressor variables,
mtry <- tuneRF(df_, df_$max_severity, ntreeTry=2000,stepFactor=1,improve=0.01, trace=TRUE, plot=FALSE, stringsAsFactors = FALSE)
best.m <- mtry[mtry[, 2] == min(mtry[, 2]), 1]
print(best.m) #is the best mtry, mtry=best.m
rf = randomForest(max_severity~.,data=df_, mtry=best.m, importance=TRUE,ntree=3000,keep.forest=TRUE,
                  sampsize=c('0'=50,'2'=50), strata=df_$max_severity, stringsAsFactors = FALSE)
