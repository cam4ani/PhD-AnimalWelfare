## Test difference

```r
path_cl = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/chapter0_final_/Cleaning'
df_transition4test = read.csv(file.path(path_cl, 'df_transition4test.csv'), header = TRUE, sep = ";")
df_transition4test$variable = as.factor(df_transition4test$variable)
df_transition4test$BatchID = as.factor(df_transition4test$BatchID)
df_transition4test$zone = as.factor(df_transition4test$zone)
df_transition4test = within(df_transition4test, variable <- relevel(variable, ref = 'Observed_Zone'))
print(dim(df_transition4test))
df_transition4test
```

```r
#Perform the chi-square goodness of fit test 
#poisson distribution to model count data
gf = goodfit(df_transition4test$value,type= "nbinomial",method= "MinChisq")
summary(gf)
plot(gf,main="Count data vs Negative binomial distribution")
```
