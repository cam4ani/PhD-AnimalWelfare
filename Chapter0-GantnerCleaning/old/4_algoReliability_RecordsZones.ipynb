# Video & Focals & Scan observations systems statistics

```r
path_records = 'R:\\VPHI\\Welfare\\2- Research Projects\\OFHE2.OriginsE2\\DataOutput\\TrackingSystem\\v3_\\Cleaning'
```

```r
#open file
df = read_delim(file.path(path_records,'v3__RecordReliabilityCleaningMethods.csv'),';',trim_ws = TRUE)

#into factor for confusion matrix and test
df[c('ObservedZone','tracking_ObservedZone','model_ObservedZone','PenID')] = lapply(df[c('ObservedZone','tracking_ObservedZone',
                                                                                     'model_ObservedZone','PenID')], factor)
```

```r
################################################################
######################## Tracking system #######################
################################################################

################ reliability
print('------------------------------------------------------------------------------------')
print('----------------------------------- RELIABILITY-------------------------------------')
print('------------------------------------------------------------------------------------')

print('-------------- all pen')
kappa2(df[,c('ObservedZone','tracking_ObservedZone')], "unweighted")
agree(df[,c('ObservedZone','tracking_ObservedZone')], tolerance=0)

#reliability for each systems separately
c_pens = list(c(3,4,5),c(8,9),c(10,11,12))
for (c_pen in c_pens){
    df_ = subset(df, df$PenID %in% c_pen)
    print(paste0('-------------- pen: ',c_pen))
    print(kappa2(df_[,c('ObservedZone','tracking_ObservedZone')], "unweighted"))
    print(agree(df_[,c('ObservedZone','tracking_ObservedZone')], tolerance=0))
    }

################ mismatch
print('------------------------------------------------------------------------------------')
print('------------------------------------- MISMATCH -------------------------------------') 
print('------------------------------------------------------------------------------------')

print('-------------- all pen')
confusionMatrix(df$tracking_ObservedZone, df$ObservedZone)

#mismatch for each systems separately
c_pens = list(c(3,4,5),c(8,9),c(10,11,12))
for (c_pen in c_pens){
    df_ = subset(df, df$PenID %in% c_pen)
    print(paste0('-------------- pen: ',c_pen))
    print(confusionMatrix(df_$tracking_ObservedZone, df_$ObservedZone))
    }        
```

```r
################################################################
############################ Model #############################
################################################################

################ reliability
print('------------------------------------------------------------------------------------')
print('----------------------------------- RELIABILITY-------------------------------------')
print('------------------------------------------------------------------------------------')

print('-------------- all pen')
kappa2(df[,c('ObservedZone','model_ObservedZone')], "unweighted")
agree(df[,c('ObservedZone','model_ObservedZone')], tolerance=0)

#reliability for each systems separately
c_pens = list(c(3,4,5),c(8,9),c(10,11,12))
for (c_pen in c_pens){
    df_ = subset(df, df$PenID %in% c_pen)
    print(paste0('-------------- pen: ',c_pen))
    print(kappa2(df_[,c('ObservedZone','model_ObservedZone')], "unweighted"))
    print(agree(df_[,c('ObservedZone','model_ObservedZone')], tolerance=0))
    }

################ mismatch
print('------------------------------------------------------------------------------------')
print('------------------------------------- MISMATCH -------------------------------------') 
print('------------------------------------------------------------------------------------')

print('-------------- all pen')
confusionMatrix(df$model_ObservedZone, df$ObservedZone)

#mismatch for each systems separately
c_pens = list(c(3,4,5),c(8,9),c(10,11,12))
for (c_pen in c_pens){
    df_ = subset(df, df$PenID %in% c_pen)
    print(paste0('-------------- pen: ',c_pen))
    print(confusionMatrix(df_$model_ObservedZone, df_$ObservedZone))
    }        
```
