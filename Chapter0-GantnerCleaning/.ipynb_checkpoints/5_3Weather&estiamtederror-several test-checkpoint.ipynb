# Download data

```r
path_ = 'G:/VPHI/Welfare/2- Research Projects/OFHE2.OriginsE2/DataOutput/TrackingSystem/CleaningAnalysis_/Cleaning/Raw&Cleaned'
df = read.csv(file.path(path_,'environmentalfactorestimatederrorrate.csv'), header = TRUE, sep = ";")
df$hour = as.integer(df$hour)   
df$PenID = as.factor(df$PenID) 
df$system = as.factor(df$system) 
#arcsine transformation in R : https://www.programmingr.com/tutorial/arcsine-transformation/
#df$estimated.error.rate.arcsine = asin(sqrt(df$estimated.error.rate))
df$humidity_avgpercent_tr = df$humidity_avgpercent/10
print(dim(df))
summary(df)
```

# Generalized linear mixed effect model

```r
fit = glmer(cbind(IFR,nbr.registrations) ~ humidity_avgpercent_tr + temperature_C + hour + estimated.transitions + (1|system/PenID), 
            family=binomial, data=df)
summary(fit)
```

```r
#https://cran.r-project.org/web/packages/glmmTMB/vignettes/model_evaluation.pdf
library(DHARMa)
res <- simulateResiduals(fit)
plot(res)
```

#Check for Overdispersion
overdisp_fun <- function(model) {
  rdf <- df.residual(model)
  rp <- residuals(model,type="pearson")
  Pearson.chisq <- sum(rp^2)
  prat <- Pearson.chisq/rdf
  pval <- pchisq(Pearson.chisq, df=rdf, lower.tail=F)
  c(chisq=Pearson.chisq, ratio=prat, rdf=rdf, p=pval)
}
overdisp_fun(fit)
#--> ratio <1 underestimation of the 0 count
#https://www.rdocumentation.org/packages/sjstats/versions/0.17.1/topics/overdisp

# linear mixed effect model

```r
#(1|tracking/PenID): penID is nested in tracking system. 
fit = lmerTest::lmer(estimated.error.rate_norm ~ humidity_avgpercent + temperature_C + hour + estimated.transitions + (1|system/PenID), 
                     data=df)
summary(fit)
```

```r
###### normally distributed residuals
#qqmath(fit, id=0.05)
qqnorm(resid(fit), main = "Residuals")
qqline(resid(fit))
```

```r
######check homogeneity of variance (residuals has constant variance)
#variance of the residuals is equal across groups. Even spread around the centered line?
plot(fit)
```

## beta with mixed effects

```r
fit = glmmTMB(estimated.error.rate_logit ~ humidity_avgpercent + temperature_C + hour + estimated.transitions + (1|system/PenID),
        df, family=beta_family(link = "logit"))
summary(fit)
```

```r
#https://cran.r-project.org/web/packages/glmmTMB/vignettes/model_evaluation.pdf
library(DHARMa)
res <- simulateResiduals(fit)
plot(res)
```
