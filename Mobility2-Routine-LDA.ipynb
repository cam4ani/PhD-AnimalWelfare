```python
path_extracted_data = config.path_extracted_data
id_run = config.id_run
nbr_sec = config.nbr_sec
path_initial_data = config.path_initial_data
#day level
title_ = '_all'
```

## Download topics distribution per hen

```python
#add all topic modelling into one dataframe
li_dflda = []
min_topic = 2
max_topic = 14
df_all_lda = pd.DataFrame(columns=['day','HenID'])
for nbr_topics_ in tqdm.tqdm(range(min_topic,max_topic+1)):
    print(df_all_lda.shape)
    path_save = os.path.join(path_extracted_data,'visual','LDA', title_, str(nbr_topics_))
    df_lda = pd.read_csv(os.path.join(path_save,'df_topics'+str(nbr_topics_)+'.csv'), sep=';',parse_dates=['day']) 
    #print(df_lda.shape)
    #df_lda.head(3)
    df_lda['topic'] = df_lda['topic'].map(lambda x: 'topic_'+str(x))
    #lets take the topic distribution / main topic of the clustering with k=6-15 over all the days of the first session
    li_sorted_days = df_lda.sort_values(['day'],ascending=True)['day'].unique()
    li_topics = [x for x in df_lda.columns if (x.startswith('topic')) & (x!='topic_info')]
    dico_topics_newtopics = {c:'k'+str(nbr_topics_)+'_'+ c for c in li_topics}
    df_lda.rename(columns=dico_topics_newtopics, inplace=True)
    df_all_lda = pd.merge(df_all_lda, df_lda[['HenID','day']+list(dico_topics_newtopics.values())], on=['HenID','day'], 
                          how='outer')
print(df_all_lda.shape)
print(len(df_all_lda['HenID'].unique()))
df_all_lda.head(3)
```

```python
##################################### choose nbr of topics and add it
for nbr_topics_plot in tqdm.tqdm(range(3,15)):
    path_extracted_visual = os.path.join(path_extracted_data, 'visual', 'ind_consistency', title_)
    #create a director if not existing
    if not os.path.exists(path_extracted_visual):
        os.makedirs(path_extracted_visual)

    li_topics = ['k'+str(nbr_topics_plot)+'_topic_'+str(i)+'_proba' for i in range(0,nbr_topics_plot)]
    print(li_topics)
    df_plot = df_all_lda.filter(['HenID','day']+li_topics, axis=1)
    display(df_plot.head(3))
    li_df = []
    for h,df_ in df_plot.groupby(['HenID']):
        df_['day_before'] = df_['day'].map(lambda x: x-dt.timedelta(days=1))
        li_day = df_['day'].unique()
        df_['is_day_before'] = df_['day_before'].map(lambda x: np.datetime64(x) in li_day)
        li_proba = [x for x in df_.columns if 'proba' in x]
        df_['prob_vector'] = df_.apply(lambda x: str([x[i] for i in li_proba]), axis=1)
        df_['prob_vector'] = df_['prob_vector'].map(lambda x: eval(x))
        df__ = df_[df_['is_day_before']].copy()
        df__['chi2distance_topic_mixture_with_day_before'] = df__.apply(lambda x: chi2_distance(df_[df_['day']==x['day_before']]['prob_vector'].values[0], 
                                                                                                 df_[df_['day']==x['day']]['prob_vector'].values[0],
                                                                                                 remove_warning=True), axis=1)
        #for verify
        #display(df__[li_topics+['day','chi2distance_topic_mixture_with_day_before']].head(3))
        li_df.append(df__[li_topics+['HenID','day','chi2distance_topic_mixture_with_day_before']])
        plt.plot(df__['day'].tolist(),df__['chi2distance_topic_mixture_with_day_before'].tolist())
    plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_perhen_quickplot'+str(nbr_topics_plot)+'.png'),dpi=300,format='png',bbox_inches='tight')
    plt.show()

    df_plot = pd.concat(li_df)
    born_date = dt.datetime(2016,10,27)-dt.timedelta(days=21*7) #2016-10-27: 21 weeks 
    df_plot['nbr_of_days'] = df_plot['day'].map(lambda x: (x-born_date).days)
    print(df_plot.shape)
    display(df_plot.head(3))


    ##################################### add info
    df_movement = pd.read_csv(os.path.join(path_extracted_data, 'df_movement.csv'), sep=';', parse_dates=['day'])
    df_movement = df_movement.filter(['HenID','SessionID','day','severity','hybrid']).drop_duplicates(subset=None, keep='first')
    print(df_movement.shape)
    display(df_movement.head(3))

    li_path = glob.glob(os.path.join(path_extracted_data, id_run+'_time_serie_day_*_variables.csv'))
    li_df = []
    for path_ in tqdm.tqdm(li_path):
        #open the sessions-variable csv and select correct hen
        df_sess = pd.read_csv(path_, sep=';',
                               parse_dates=['level','FirstTimestamp_zone_1','FirstTimestamp_zone_2', 'FirstTimestamp_zone_3', 
                                            'FirstTimestamp_zone_4', 'FirstTimestamp_zone_5'])
        df_sess.rename(columns={'level':'day'}, inplace=True)
        li_df.append(df_sess)
    df_henLevel_summary = pd.concat(li_df)
    print(df_henLevel_summary.shape)
    display(df_henLevel_summary.head(3))

    df_all = pd.merge(df_henLevel_summary, df_movement, on=['HenID','day'], how='left')
    print(df_all.shape)
    df_all.head(3)

    print(df_plot.shape)
    df_plot = pd.merge(df_plot, df_all, on=['HenID', 'day'], how='left')
    print(df_plot.shape)
    df_plot.head(2)

    df_sev = df_plot.groupby(['HenID', 'SessionID'])['severity'].agg(lambda x: list(x)[0] if len(set(x))==1 else -2).reset_index()
    print(df_sev[df_sev['severity']==-2].shape)
    df_sev = df_sev.groupby(['HenID'])['severity'].max().reset_index()
    df_sev.rename(columns={'severity':'max_severity'}, inplace=True)
    df_plot = pd.merge(df_plot, df_sev, on=['HenID'], how='left')
    df_plot['severity'].fillna(-1, inplace=True)
    df_plot['Total_number_transition_cat'] = df_plot['Total_number_transition'].map(lambda x: x-x%30)
    #df_plot['reasonable_KF'] = df_plot['max_severity'].map(lambda x: x<6)
    df_plot['max_severity'] = df_plot['max_severity'].map(lambda x: round(x))
    df_plot['round_severity'] = df_plot['severity'].map(lambda x: round(x))
    df_plot['round_severity'] = df_plot['severity'].map(lambda x: round(x))
    df_plot['reasonable_KF'] = df_plot['max_severity'].map(lambda x: x<8)
    print(df_plot.shape)
    df_plot.head(2)
    print(len(df_plot['HenID'].unique()))


    ##################################### plot 
    # robust=True, markers=["o", "x"]; # order=2
    g = sns.lmplot(x="nbr_of_days", y="chi2distance_topic_mixture_with_day_before", hue="HenID", data=df_plot, size=10, palette="Set1", legend=False); 
    plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_perhen_'+str(nbr_topics_plot)+'.png'),dpi=300,format='png',  bbox_inches='tight')

    g = sns.lmplot(x="nbr_of_days", y="chi2distance_topic_mixture_with_day_before", data=df_plot, size=10, palette="Set1",  legend=False, lowess=True);
    plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_allhen_'+str(nbr_topics_plot)+'.png'),dpi=300,format='png',bbox_inches='tight')

    g = sns.lmplot(x="nbr_of_days", y="chi2distance_topic_mixture_with_day_before", hue="max_severity", data=df_plot, size=10, 
                   palette=sns.color_palette("RdBu_r", n_colors=len(df_plot['max_severity'].unique()))); 
    plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_permaxseverity_'+str(nbr_topics_plot)+'.png'),dpi=300,format='png',bbox_inches='tight')

    g = sns.lmplot(x="nbr_of_days", y="chi2distance_topic_mixture_with_day_before", hue="reasonable_KF", data=df_plot, size=10); 
    plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_perreasonableKF_'+str(nbr_topics_plot)+'.png'),dpi=300,format='png', 
                bbox_inches='tight')

    if title_=='_all':
        g = sns.lmplot(x="nbr_of_days", y="chi2distance_topic_mixture_with_day_before", hue="hybrid", data=df_plot, size=10); 
        plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_perhybrid_'+str(nbr_topics_plot)+'.png'),dpi=300,format='png', 
                    bbox_inches='tight')
        
        df_plot_LSL = df_plot[df_plot['hybrid']=='LSL']
        g = sns.lmplot(x="nbr_of_days", y="chi2distance_topic_mixture_with_day_before", hue="max_severity", data=df_plot_LSL, size=10, 
                       palette=sns.color_palette("RdBu_r", n_colors=len(df_plot_LSL['max_severity'].unique()))); 
        plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_permaxseverity_LSL_'+str(nbr_topics_plot)+'.png'),dpi=300,format='png',bbox_inches='tight')
        
        df_plot_LB = df_plot[df_plot['hybrid']=='LB']
        g = sns.lmplot(x="nbr_of_days", y="chi2distance_topic_mixture_with_day_before", hue="max_severity", data=df_plot_LB, size=10, 
                       palette=sns.color_palette("RdBu_r", n_colors=len(df_plot_LB['max_severity'].unique())));
        plt.savefig(os.path.join(path_extracted_visual, 'chi2distance_LDA_overtime_permaxseverity_LB_'+str(nbr_topics_plot)+'.png'),dpi=300,format='png',bbox_inches='tight')        
        
        #lot to see how it separates the hybrids in the topics
        for h in df_plot[df_plot['hybrid']=='LB']['HenID'].unique():
            df__ = df_plot[(df_plot['HenID']==h)]
            #plt.plot(df__['day'].tolist(), [i/10 for i in df__['severity'].tolist()], color='red');
            df__[li_topics].plot(x=df__.day, kind='bar', stacked=True, legend=True)
            plt.title(h+'  max_severity: '+str(df__['max_severity'].values[0]))
            plt.savefig(os.path.join(path_extracted_data, 'visual', 'LDA', title_, str(nbr_topics_plot), 'topics distribution_k'+str(nbr_topics_plot)+'LB_hen_'+h+'.png'),dpi=300,format='png',bbox_inches='tight')        
            plt.clf()
        for h in df_plot[df_plot['hybrid']=='LSL']['HenID'].unique():
            df__ = df_plot[(df_plot['HenID']==h)]
            #plt.plot(df__['day'].tolist(), [i/10 for i in df__['severity'].tolist()], color='red');
            df__[li_topics].plot(x=df__.day, kind='bar', stacked=True, legend=True)
            plt.title(h+'  max_severity: '+str(df__['max_severity'].values[0]))
            plt.savefig(os.path.join(path_extracted_data, 'visual', 'LDA', title_, str(nbr_topics_plot), 'topics distribution_k'+str(nbr_topics_plot)+'LSL_hen_'+h+'.png'),dpi=300,format='png',bbox_inches='tight')        
            plt.clf()
    plt.clf()
#--> when you are not in pain your routine get stable over time. In contrary, when you have a high keel born fracture they
#dont seemed able to do their normal routine?
```
