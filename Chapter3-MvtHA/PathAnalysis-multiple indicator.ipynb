```r
#https://www.rdocumentation.org/packages/lcsm/versions/0.1.2/topics/fit_bi_lcsm
# Fit bivariate lcsm and save the results 
bi_lavaan_results <- fit_bi_lcsm(data = df, 
                                 var_y = c("vertical_travel_distance_HA2", "vertical_travel_distance_HA4", "vertical_travel_distance_HA5"),
                                 var_x = c("severity_HA2", "severity_HA4", "severity_HA5"),
                                 model_x = list(alpha_constant = TRUE, 
                                                beta = TRUE,  #not sure!!
                                                phi = FALSE),
                                 model_y = list(alpha_constant = TRUE, 
                                                beta = TRUE, 
                                                phi = TRUE),
                                 coupling = list(delta_lag_xy = TRUE, 
                                                 xi_con_yx = TRUE))

# Save the lavaan syntax that is used to create the layout matrix for semPlot
bi_lavaan_syntax <- fit_bi_lcsm(data = df, 
                                 var_y = c("vertical_travel_distance_HA2", "vertical_travel_distance_HA4", "vertical_travel_distance_HA5"),
                                 var_x = c("severity_HA2", "severity_HA4", "severity_HA5"),
                                model_x = list(alpha_constant = TRUE, 
                                               beta = TRUE, 
                                               phi = FALSE),
                                model_y = list(alpha_constant = TRUE, 
                                               beta = TRUE, 
                                               phi = TRUE),
                                 coupling = list(delta_lag_xy = TRUE, 
                                                 xi_con_yx = TRUE),
                                return_lavaan_syntax = TRUE,
                                change_letter_x = "M",
                                change_letter_y = "H")

#result
summary(bi_lavaan_results)

# Plot the results
plot_lcsm(lavaan_object = bi_lavaan_results, 
          lavaan_syntax = bi_lavaan_syntax,
          lcsm_colours = TRUE,
          lcsm = "bivariate")
```
