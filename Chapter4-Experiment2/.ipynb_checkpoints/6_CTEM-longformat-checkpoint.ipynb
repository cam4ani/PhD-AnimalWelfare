# Hierarchical Bayesian continuous time dynamic model - VTD & KBF severity

```r
#initialise what we want
name = 'VTD_severity'
path_save_ = file.path(path_save,'VTD_severity')
dir.create(path_save_)
```

```r
#processing dataframe to match expected format
data_VTDsev = df %>% select (HenID, time, severity, VTDperhour) #or DOAAVG instead of time
data_VTDsev = plyr::rename (data_VTDsev, c("HenID"="id", "severity"="Y1", "VTDperhour"="Y2"))
summary(data_VTDsev)
```

```r
#centering and scaling variables in the model
data_VTDsev[,c('Y1','Y2')] <- scale(data_VTDsev[,c('Y1','Y2')])
summary(data_VTDsev)
str(data_VTDsev)
```

```r
#initialize the model with two latent each with ne manifest variable
set.seed(1)
model_fit = ctModel(type='stanct', latentNames=c('eta1','eta2'), manifestNames=c('Y1','Y2'), LAMBDA=diag(2)) 
#set a narrow prior for the random scale #TODO: check sensitivity to other ones
if(TRUE){
    model_fit$pars$indvarying = TRUE
    model_fit$pars$sdscale[!model_fit$pars$matrix %in% c('T0MEANS','MANIFESTMEANS','CINT')] = 0.05
}

#check model parameters
print(model_fit$pars)
```

```r
#fit the model
fit_VTDseverity = ctStanFit(datalong=data_VTDsev, ctstanmodel=model_fit, plot=10, verbose=0, cores=cores, nopriors=F)

#print the summary
#with: folder=path_save_ impossible de changer de répertoire de travail
ctsem:::ctSummarise(fit_VTDseverity, folder=name, cores=cores, lags=c(1,2,3,5,10,15,20,25,30,35),
                    ctStanPlotPost=F, nsamples=1000)

#summary of the model
#calculate discrete time parameters (in weeks)
summary(fit_VTDseverity, timeinterval=1)
summary(fit_VTDseverity, timeinterval=2)
summary(fit_VTDseverity, timeinterval=3)
summary(fit_VTDseverity, timeinterval=4)

plot(fit_VTDseverity)

#plot the whole model & check model convergence
ctStanDiscretePars(fit_VTDseverity, plot=T)
```

```r
fit_VTDseverity$setup$idmap[c(1,100,200,436), 1]
```

```r
#plot model obs for 5 random hens (max number is I think 436)
ctKalman(fit_VTDseverity,subjects=fit_VTDseverity$setup$idmap[c(1,100,200,300,400), 1],plot=T,kalmanvec=c('y','ysmooth'),
         polygonsteps=0)
```

```r
#plot model obs for 5 random hens (max number is I think 436)
ctKalman(fit_VTDseverity,subjects=fit_VTDseverity$setup$idmap[c(10,110,210,310,410), 1],plot=T,kalmanvec=c('y','ysmooth'),
         polygonsteps=0)
```

```r
#plot parameters for a single hen (the second one, as the first one do not work)
ctKalman(fit_VTDseverity,subjects=fit_VTDseverity$setup$idmap[1, 1],plot=T,kalmanvec=c('y','ysmooth'))
```

```r
#extract ind. parameters
indpars = ctStanSubjectPars(fit_VTDseverity, cores=cores, nsamples="all")[1,,]
indpars
str(indpars)
dimnames(indpars)

#convert to a data-frame
ind.data = data.frame(indpars)
summary(ind.data)

#check IDs
print(data_VTDsev$id)
fit_VTDseverity$setup$idmap
id = data.frame(levels(data_VTDsev$id))
colnames(id)
ind.data$ID = as.factor(id$levels.data_VTDsev.id.)
str(ind.data)

#write a csv
write.csv(ind.data, file=file.path(path_save_,"HenIDParametersVTDandSeverity.csv"))
```

```r
#additional plot
ctfit_VTDseverity = read.table(file.path(path_save_,"HenIDParametersVTDandSeverity.csv"), 
                               header=T, sep = ",", na.strings = "NA")
ctfit_VTDseverity$ID = as.factor(ctfit_VTDseverity$ID)
summary(ctfit_VTDseverity)


#GRAPHS####
par(mfrow=c(1,2))
names(ctfit_VTDseverity)

#Model 1: SP frequency and Well-being
#Distribution of continuous individual auto-effects
hist(ctfit_VTDseverity$drift_eta1,breaks=5,
     xlim=range (-1, 1), ylim=range(0, 35),freq=T,
     xlab= "KBF severity", main=NA,ylab="n of participants")

hist(ctfit_VTDseverity$drift_eta2,breaks=5,
     xlim=range (-4,1), ylim=range(0, 35),freq=T,
     xlab= "VTD ", main=NA,ylab="n of participants")


#Distribution of continuous individual cross-lagged effects
hist(ctfit_VTDseverity$drift_eta1_eta2,
     breaks=5,
     xlim=range (-.5, .5), freq=T,ylim=range(0, 35),
     xlab= "VTD -->  KBF severity", ylab="n of participants",
     main = NA)

hist(ctfit_VTDseverity$drift_eta2_eta1,
     breaks=5,
     xlim=range (-.5, .5), ylim=range(0, 35),freq=T,
     xlab= "KBF severity --> VTD", main=NA,ylab="n of participants")
```
