```python
#search for the repeatability csv files, download each file and concatenate them
li_df = []
for path_ in glob.glob(os.path.join(path_extracted_data,'OFH_R_estimates_*.csv')):
    print('--------------------- files %s'%path_.split('\\')[-1])
    df_ = pd.read_csv(path_, sep=',', index_col=0)
    print(df_.shape) 
    display(df_.head(3))
    li_df.append(df_)
print('------------------------- ALL TOGETHER:')
df_R = pd.concat(li_df)  
df_R['VI_LL'] = df_R['li_VI'].map(lambda x: float(x.split(', ')[0].split('[')[1]))
df_R['VI_UL'] = df_R['li_VI'].map(lambda x: float(x.split(', ')[1].strip(']')))
df_R['VE_LL'] = df_R['li_VE'].map(lambda x: float(x.split(', ')[0].split('[')[1]))
df_R['VE_UL'] = df_R['li_VE'].map(lambda x: float(x.split(', ')[1].strip(']')))
df_R['VI'] = df_R['li_VI'].map(lambda x:float(x.split(' ')[0]))
df_R['VE'] = df_R['li_VE'].map(lambda x:float(x.split(' ')[0]))
display(df_R[['li_VI','li_VE','VI','VE','VI_LL','VI_UL','VE_LL','VE_UL']].head(3))
df_R['li_mvt_named'] = df_R['li_mvt'].map(lambda x: dico_mvt_name[x])
print(df_R.shape) 
display(df_R.head(3))
```

```python
#add nice text to output in the paper (i.e. with specific number of values after the comma, adding useless 0)
#add R and 95% interval as a text column
df_R['text'] = df_R.apply(lambda x: ("%.2f" % x['li_R'])+' ['+("%.2f" % x['li_CIL'])+', '+("%.2f" % x['li_CIU'])+']', 
                          axis=1)
df_R['li_VI'] = df_R.apply(lambda x: ("%.4f" % x['VI'])+' ['+("%.4f" % x['VI_LL'])+', '+("%.4f" % x['VI_UL'])+']', 
                          axis=1)
df_R['li_VE'] = df_R.apply(lambda x: ("%.4f" % x['VE'])+' ['+("%.4f" % x['VE_LL'])+', '+("%.4f" % x['VE_UL'])+']', 
                          axis=1)
df_R.head(3)
```

```python
df_R['VI_LL_'] = df_R.apply(lambda x: x['VI']-x['VI_LL'], axis=1)
df_R['VE_LL_'] = df_R.apply(lambda x: x['VE']-x['VE_LL'], axis=1)
df_R['VI_UL_'] = df_R.apply(lambda x: x['VI_UL']-x['VI'], axis=1)
df_R['VE_UL_'] = df_R.apply(lambda x: x['VE_UL']-x['VE'], axis=1)
display(df_R[['VI','VI_LL','VI_UL','VI_LL_','VI_UL_']].head(2))
df_R['li_R_IC'] = df_R.apply(lambda x: (x['li_R'],x['li_CIL_'],x['li_CIU_']), axis=1)
df_R['li_VI_IC'] = df_R.apply(lambda x: (x['VI'],x['VI_LL_'],x['VI_UL_']), axis=1)
df_R['li_VE_IC'] = df_R.apply(lambda x: (x['VE'],x['VE_LL_'],x['VE_UL_']), axis=1)
df_plt = pd.melt(df_R, id_vars=['li_mvt','li_mvt_','li_BStype','li_mvt_named'], value_vars=['li_R_IC', 'li_VI_IC', 'li_VE_IC'], value_name = 'tuple')
df_plt['value'] = df_plt['tuple'].map(lambda x: x[0])
df_plt['CIL'] = df_plt['tuple'].map(lambda x: x[1])
df_plt['CIU'] = df_plt['tuple'].map(lambda x: x[2])
df_plt['type'] = df_plt.apply(lambda x: x['li_mvt_']+x['variable'], axis=1)
dico_var_order = {'li_R_IC':3,'li_VI_IC':1,'li_VE_IC':0} #Add treatment!
df_plt['order'] = df_plt['variable'].map(lambda x: dico_var_order[x])
df_plt = df_plt.sort_values(['order','li_BStype'], ascending=True)
display(df_plt.head(3))
for mvt in df_R['li_mvt'].unique():
    plt.figure(figsize=(3,3))
    df_plt__ = df_plt[(df_plt['li_mvt']==mvt)].copy()
    df_plt__.loc[~df_plt__['variable'].isin(['li_VI_IC','li_VE_IC']), 'value'] = np.nan
    df_plt__.loc[~df_plt__['variable'].isin(['li_VI_IC','li_VE_IC']), 'CIU'] = np.nan
    df_plt__.loc[~df_plt__['variable'].isin(['li_VI_IC','li_VE_IC']), 'CIL'] = np.nan
    sns.scatterplot(x='type', y='value', data=df_plt__, hue='li_BStype', palette=config.pal_treat, s=80);
    plt.errorbar(df_plt__['type'].tolist(), df_plt__['value'].tolist(),
                 yerr=[df_plt__['CIL'].tolist(),df_plt__['CIU'].tolist()], fmt="none", alpha=0.4, ms=5, color='black');
    df_plt__['CIL_'] = df_plt__.apply(lambda x: x['value']-x['CIL'], axis=1)
    df_plt__['CIU_'] = df_plt__.apply(lambda x: x['value']+x['CIU'], axis=1)
    plt.ylim(df_plt__['CIL_'].min()*0.9, df_plt__['CIU_'].max()*1.1)
    plt.xticks(rotation=90);
    plt.ylabel('Variance')
    plt.grid(True)

    ax2 = plt.twinx()
    df_plt__ = df_plt[(df_plt['li_mvt']==mvt)].copy()
    df_plt__.loc[df_plt__['variable'].isin(['li_VI_IC','li_VE_IC']), 'value'] = np.nan
    sns.scatterplot(x='type', y='value', data=df_plt__, hue='li_BStype', palette=config.pal_treat, ax=ax2, legend=False,
                   s=80);
    plt.errorbar(df_plt__['type'].tolist(), df_plt__['value'].tolist(),
                 yerr=[df_plt__['CIL'].tolist(),df_plt__['CIU'].tolist()], fmt='none', alpha=0.4,ms=5, color='black');    
    plt.ylim(0, 1)
    plt.xticks(rotation=90);
    plt.ylabel('R')
    plt.title(mvt)
    plt.savefig(os.path.join(path_extracted_data,'OFH_vriance_'+mvt+'.png'), bbox_inches='tight')
    plt.show();
```
