# KBF severity

```r
dim(df)
df_S = df[(!is.na(df$severity))&(df$DOA>200),]
dim(df_S)
```

```r
hist(df_S$severity)
hist(sqrt(df_S$severity))
```

#worst without sqrt
#penID as random effect: singular values
fit = lmer(severity ~  CLASS +date*Treatment+(1|HenID), data=df_S)
summary(fit)
######normally distributed residuals
qqnorm(resid(fit))
qqline(resid(fit))
hist(resid(fit))
######check homogeneity of variance (residuals has constant variance)
plot(fit)
#isnt it comapred to the ref group?

```r
#sqrt is better. penid no variance
#penID as random effect: singular values
fit = lmer(sqrt(severity) ~  CLASS + DOA+dateTreatment+(1|PenID/HenID), data=df_S)
summary(fit)
######normally distributed residuals
qqnorm(resid(fit))
qqline(resid(fit))
hist(resid(fit))
######check homogeneity of variance (residuals has constant variance)
plot(fit)
#isnt it comapred to the ref group?
```

```r
#sqrt is better. penid no variance: reoved it
#penID as random effect: singular values
fit = lmer(sqrt(severity) ~  CLASS + dateTreatment+(1|HenID), data=df_S)
summary(fit)
######normally distributed residuals
qqnorm(resid(fit))
qqline(resid(fit))
hist(resid(fit))
######check homogeneity of variance (residuals has constant variance)
plot(fit)
#isnt it comapred to the ref group?
```

```r
#penID as random effect: singular values -->withouPenID
fit = lmer(sqrt(severity) ~  CLASS +DOA+dateTreatment + (1|HenID), data=df_S)
summary(fit)
anova(fit)
######normally distributed residuals
qqnorm(resid(fit))
qqline(resid(fit))
hist(resid(fit))
######check homogeneity of variance (residuals has constant variance)
plot(fit)

emm1 = emmeans(fit, specs = ~ dateTreatment)
emm1
df_c = as.data.frame(contrasts(df_S$dateTreatment))
df_c

OFH215 = c(1,0,0,0,0,0,0,0)
TRAN215 = df_c[,c('215.TRAN')]

OFH243 = df_c[,c('243.OFH')]
TRAN243 = df_c[,c('243.TRAN')]

OFH313 = df_c[,c('313.OFH')]
TRAN313 = df_c[,c('313.TRAN')]

OFH417 = df_c[,c('417.OFH')]
TRAN417 = df_c[,c('417.TRAN')]

contrast(emm1, method = list('215.OFH-215.TRAN'=OFH215-TRAN215,
                             '243.OFH-243.TRAN'=OFH243-TRAN243, 
                             '313.OFH-313.TRAN'=OFH313-TRAN313, 
                             '417.OFH-417.TRAN'=OFH417-TRAN417), adjust='bonferroni')
```
