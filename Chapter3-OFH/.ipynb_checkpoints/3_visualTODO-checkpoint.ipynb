```python
li_p = glob.glob(os.path.join(path_extracted_data, 'Pred_fit*.csv'))
li_df = []
for p in li_p:
    print(p)
    df = pd.read_csv(p, sep=',', index_col=0).reset_index()
    df['name'] = p.split('\\')[-1]
    li_df.append(df)
    print(df.shape)
    display(df.head(3))
df = pd.concat(li_df)
df['y_errUP'] = df.apply(lambda x: x['conf.high']-x['predicted'], axis=1)
df['y_errDO'] = df.apply(lambda x: x['predicted']-x['conf.low'], axis=1)
display(df.head(3))
print(df.shape)
```

```python
df['name'].unique()
```

```python
def class_fct(x):
    if 'fit2' in x:
        return('proportion_indoor_time_in_litter_floor')
    elif 'fitvtd' in x:
        return('vertical_travelled_distance')
    elif 'fitwg' in x:
        return('number_of_visit_to_WG')
    elif 'fitZ4Duration' in x:
        return('proportion_indoor_time_in_nestbox_tier')
    elif 'fit5' in x:
        return('proportion_indoor_time_in_top_tier')
    elif 'fit4M' in x:
        return('mid-nestboox_tier_usage')
    else:
        return('error')
df['mvtvar'] = df['name'].map(lambda x: class_fct(x))
df['mvtvar'].value_counts()
```

```python
def class_fct(x):
    if 'ns3' in x:
        return('ns3')
    elif 'knot' in x:
        return('knot')
    else:
        return('poly2')
df['model_type'] = df['name'].map(lambda x:class_fct(x))
df.groupby(['model_type'])['name'].agg(lambda x: len(set(x))).reset_index()
```

```python
df_mvt = pd.read_csv(os.path.join(path_extracted_data, 'OFH_df_spaceUse_allday_4visual.csv'), sep=',', index_col=0)
df_mvt.rename(columns={'propindoor_duration_5_Zone_rs':'proportion_indoor_time_in_top_tier', 
                'propindoor_duration_4_Zone_rs':'proportion_indoor_time_in_nestbox_tier', 
                'mid_cum_Z4_h_Morning':'mid-nestboox_tier_usage',
                'nbr_stays_1_Zone':'number_of_visit_to_WG', 
                'vertical_travel_distance':'vertical_travelled_distance', 
                'propindoor_duration_2_Zone_rs':'proportion_indoor_time_in_litter_floor'}, inplace=True)
print(df_mvt.shape) 
display(df_mvt.head(3))
```

```python
for (mvt_,model_type),df_plt in df.groupby(['mvtvar','model_type']):
    fig, ax = plt.subplots()        
    for gr_, df_plt_ in df_plt.groupby('group'):
        ax.plot(df_plt_['x'].tolist(), df_plt_['predicted'].tolist(), '-', color=pal_tr[gr_])
        ax.fill_between(df_plt_['x'].tolist(), np.array(df_plt_['predicted'].tolist()) - np.array(df_plt_['y_errDO'].tolist()), 
                        np.array(df_plt_['predicted'].tolist())+np.array(df_plt_['y_errUP'].tolist()), alpha=0.2, color=pal_tr[gr_]);
    #no need constrain on this plot
    sns.boxplot(x='DIB',y=mvt_, hue='Treatment', palette=pal_tr, data=df_mvt, fliersize=0, linewidth=0.1);
    plt.title(model_type)
    plt.show();
```

```python
#do not show values outside of 5-95percentile
for (mvt_,model_type),df_plt in df.groupby(['mvtvar','model_type']):
    fig, ax = plt.subplots()  
    if mvt_=='mid-nestboox_tier_usage':
        ax.set_ylim(2, df_mvt[mvt_].max())
    if mvt_ in ['number_of_visit_to_WG','proportion_indoor_time_in_nestbox_tier']:
        ax.set_ylim(df_mvt[mvt_].min(), np.percentile(df_mvt[~df_mvt[mvt_].isnull()][mvt_],95))
    for gr_, df_plt_ in df_plt.groupby('group'):
        ax.plot(df_plt_['x'].tolist(), df_plt_['predicted'].tolist(), '-', color=pal_tr[gr_])
        ax.fill_between(df_plt_['x'].tolist(), np.array(df_plt_['predicted'].tolist()) - np.array(df_plt_['y_errDO'].tolist()), 
                        np.array(df_plt_['predicted'].tolist())+np.array(df_plt_['y_errUP'].tolist()), alpha=0.2, color=pal_tr[gr_]);
    sns.boxplot(x='DIB',y=mvt_, hue='Treatment', palette=pal_tr, data=df_mvt, fliersize=0, linewidth=0);
    plt.title(model_type)
    plt.savefig(os.path.join(path_extracted_data,name+mvt_+'boxplotDIB.png'),transparent=True, bbox_inches='tight')
    plt.show();
```
