```python
df = pd.read_csv(os.path.join(path_records, 'VideoObservation', 'VideoObservationAllStudents.csv'), sep=';') #,
                #parse_dates=['Date']) 
#df['Date_Heure'] = df['Date_Heure'].map(lambda x: dt.datetime.strptime(x, '%d.%m.%Y %H:%M:%S'))
df['Timestamp'] = pd.to_datetime(df['Date'] + ' ' + df['Heure'])
#df = df[df['VideoAnalyse']!=-1] #cannot remove now otherwise the duration might be wrong
df = df[df['worker']!='nabila']
print(df.shape)
df.head(3)
```

## Percentage of duration associated to a wrong zone due to missed/added records from the tracking system (i.e. not due to delay)

```python
#add as tracking system the end point and beging point of a batch where a chicken did not moved
df['Zone'] = df['trackZone'].copy()
df['Zone'] = np.where(df['VideoAnalyse']==3, df['Obszone_whenWrong'], df['Zone'])
```

```python
li_df = []
#more efficient to do it per hen, as it wont need to search in the whole dataframe, and we can simply shift the timestamp column
for i, df_hen in tqdm.tqdm(df.groupby(['BirdID','batch','worker'])):
    #as the next record date (sort by date, then simply shift by one row and add nan at then end)
    df_hen = df_hen.sort_values(['Timestamp'], ascending=True)
    df_hen['next_record_date'] = df_hen['Timestamp'].tolist()[1:]+[np.nan]
    #compute duration
    df_hen['duration'] = df_hen.apply(lambda x: x['next_record_date']-x['Timestamp'], axis=1)
    #add next record for the impossible movement
    df_hen['next_zone'] = df_hen['Zone'].tolist()[1:]+[np.nan]
    li_df.append(df_hen)
#put again in one dataframe
df = pd.concat(li_df)
df.head(3)
```

```python
df_correct_dur = df[df['VideoAnalyse'].isin([1,3])].groupby(['Pen','BirdID', 'batch','worker'])['duration'].sum().reset_index()
df_correct_dur.rename(columns={'duration':'correct_duration'}, inplace=True)
df_incorrect_dur = df[df['VideoAnalyse']==0].groupby(['Pen','BirdID','batch','worker'])['duration'].sum().reset_index()
df_incorrect_dur.rename(columns={'duration':'incorrect_duration'}, inplace=True)
df_correct_dur.head(3)
```

```python
df_dur = pd.merge(df_correct_dur, df_incorrect_dur, on=['Pen', 'BirdID', 'batch','worker'], how='outer')
print(df_dur.shape)
df_dur['incorrect_duration'] = df_dur['incorrect_duration'].fillna(pd.Timedelta(seconds=0))
df_dur['correct_duration'] = df_dur['correct_duration'].fillna(pd.Timedelta(seconds=0))
df_dur['percentage_correct_durations'] = df_dur.apply(lambda x: x['correct_duration']/(x['correct_duration']+x['incorrect_duration']),
                                                    axis=1)
df_dur.head(3)
```

```python
df_dur_summarize = df_dur.groupby(['Pen','BirdID','batch'])['percentage_correct_durations'].agg(lambda x: list(x)).reset_index()
df_dur_summarize['percentage_correct_durations_avgPerWorker'] = df_dur_summarize['percentage_correct_durations'].map(lambda x: np.mean(x))
df_dur_summarize
```

```python
#multiple boxplot
sns.boxplot(x='Pen', y='percentage_correct_durations_avgPerWorker', data=df_dur_summarize, palette="PRGn") #hue="sex",
sns.despine(offset=10, trim=True)
plt.title('Percentage of duration that is correct');
plt.savefig(os.path.join(path_analysis, 'percentage_correct_durations_per_pen_avgPerWorker.png'),dpi=300,format='png')
```

## Signal strength

```python
#lets count each entry only one time if the students agree on the output, and not if the students did not agree on the output
df_signstr = df[df['VideoAnalyse'].isin([0,1])].groupby(['BirdID','Pen','Timestamp'])['worker','VideoAnalyse',
                                                                         'signalstrength','trackZone'].agg(lambda x: list(x)).reset_index()
df_signstr['ss_variance'] = df_signstr['VideoAnalyse'].map(lambda x: np.var(x))
df_signstr = df_signstr[df_signstr['ss_variance']==0]
df_signstr['VideoAnalyse'] = df_signstr['VideoAnalyse'].map(lambda x: x[0])
df_signstr['signalstrength'] = df_signstr['signalstrength'].map(lambda x: x[0])
df_signstr['trackZone'] = df_signstr['trackZone'].map(lambda x: x[0])
#df_signstr.groupby(['Pen','trackZone','VideoAnalyse'])['signalstrength'].agg(lambda x)
df_signstr.head(2)
```

```python
ax = sns.violinplot(x="trackZone", y="signalstrength", hue="VideoAnalyse",data=df_signstr, palette="muted", split=True)
plt.title('Signal Strength of all pens')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.savefig(os.path.join(path_analysis, 'Signal Strength of all pens.png'),dpi=300,format='png')
plt.show()
```

```python
for p,df_ in df_signstr.groupby('Pen'):
    if len(df_['VideoAnalyse'].unique())==2:
        ax = sns.violinplot(x="trackZone", y="signalstrength", hue="VideoAnalyse",data=df_, palette="muted", split=True)
        plt.title(p)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
        plt.savefig(os.path.join(path_analysis, 'Signal Strength pen'+str(p)+'.png'),dpi=300,format='png')
        plt.show()
    else:
        ax = sns.violinplot(x="trackZone", y="signalstrength", hue="VideoAnalyse",data=df_, palette="muted")
        plt.title(p)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
        plt.savefig(os.path.join(path_analysis, 'Signal Strength pen'+str(p)+'.png'),dpi=300,format='png')
        plt.show()
```
