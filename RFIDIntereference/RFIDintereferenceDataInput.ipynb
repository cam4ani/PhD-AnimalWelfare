```python
path_ = r'G:\VPHI\Welfare\2- Research Projects\OFHE2.OriginsE2\GantnerSystem\RFID_Interference'
df_records = pd.read_csv(os.path.join(path_,'RFID_VIDEO_result_ALL.csv'), parse_dates=['Date'], dayfirst=True, sep=';')
df_records = df_records.dropna(subset=['VideoAnalyse'])
df_records['Timestamp'] = df_records.apply(lambda x: dt.datetime(x['Date'].year, x['Date'].month, x['Date'].day, 
                                                             int(x['Time'].split(':')[0]), int(x['Time'].split(':')[1]), 
                                                             int(x['Time'].split(':')[2])), axis=1)
df_records.drop_duplicates(subset=['Timestamp','TagID'], keep='last', inplace=True)
print(df_records.shape)
df_records = df_records[df_records['VideoAnalyse']!='-1']
#RFID on: 11h-13h on the 9th and off from 13h to 15h : 4 animals per pen on the times
#RFID on: 13h-15h on the 15th and off from 11h-13h on the 16th: 4 animals per pen on those times
display(df_records['RFID_ON'].value_counts())
df_records['type'] = 'students'
print(df_records.shape)
df_records.head(3)
```

```python
#first check the accuracy per verification batch
df_ver = df_records.groupby(['BatchID','RFID_ON'])[['VideoAnalyse','Zone']].agg(lambda x: list(x)).reset_index()
df_ver['VideoAnalyse'] = df_ver['VideoAnalyse'].map(lambda x: [i for i in x if i in ['0','1']])
df_ver['accuracy'] = df_ver['VideoAnalyse'].map(lambda x: sum([i=='1' for i in x])/len(x))
df_ver['nbr_transition'] = df_ver['VideoAnalyse'].map(lambda x: len(x))
display(df_ver[df_ver['accuracy']==min(df_ver['accuracy'].tolist())])
df_ver.head(5)
```

# Add henID, previous and next zones from raw data

```python
#RFID on: 11h-13h on the 9th and off from 13h to 15h : 4 animals per pen on the times
#RFID on: 13h-15h on the 15th and off from 11h-13h on the 16th: 4 animals per pen on those times
#for being more reproductible, we open the file that was saved from cleaning
df = pd.read_csv(os.path.join(path_extracted_data, id_run+'_Rawdata&Model.csv'), sep=';', parse_dates=['Timestamp','date'],
                dayfirst=True) 
print(df[df['model_prediction'].isnull()].shape)
print(df.shape)
df = df[((df['Timestamp']>=dt.datetime(2021,2,9,11,0,0))&(df['Timestamp']<=dt.datetime(2021,2,9,15,10,0)))|\
        ((df['Timestamp']>=dt.datetime(2021,2,15,13,0,0))&(df['Timestamp']<=dt.datetime(2021,2,15,15,10,0)))|\
        ((df['Timestamp']>=dt.datetime(2021,2,16,11,0,0))&(df['Timestamp']<=dt.datetime(2021,2,16,13,10,0)))]
df = df[df['PenID'].isin(['pen10','pen11','pen12'])]
display(df['PenID'].value_counts())
print(df.shape)
df.head(3)
```
