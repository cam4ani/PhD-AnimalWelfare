```python
#basic package
import json
import tqdm
import os
import numpy as np
import glob
import pandas as pd
import sys
import shutil
import time
import datetime as dt
import itertools
import time
import re
from collections import Counter
from collections import defaultdict
import operator
import math
import functools
import collections

#plot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_ = os.path.join(r'D:\vm_exchange\PhD_Data\Productivity')
```

```python
li_df = []
for d in [str(i)+'_4_E219' for i in range(1,12)]:
    print(d)
    for v in ['('+str(i)+')' for i in range(1,21)]:
        xls = pd.ExcelFile(glob.glob(os.path.join(path_,d+'.xls*'))[0])
        df = pd.read_excel(xls, v, header=None)
        df = df.iloc[2:,]
        df.columns = df.iloc[0]
        df = df.drop(2, axis=0)
        df = df[~df['Vers.-Tag'].isnull()] #remove all rows taht has nothing to do with the egg per day
        df['LayingPhase'] = d
        df['Pen'] = int(v[1:-1])
        if df.shape[0]!=28:
            print('ERROR: ', d, v,' have wrong shape', df.shape)
        if df.shape[1]!=15:
            print('ERROR: ', d, v,' have wrong shape', df.shape)
        df.head(2)
        li_df.append(df)
        df_test = pd.concat(li_df)
df = pd.concat(li_df)
df['Nest-eier'].fillna(0,inplace=True)
df['Tier-bestand'].fillna(0,inplace=True)
df['#eggPerTier'] = df.apply(lambda x: x['Nest-eier']/x['Tier-bestand'],axis=1)
#9 septembre 2020: 420 DOA --> ne le ~ 9 juillet 2019
df['#eggOutsideBoxPerTier'] = df.apply(lambda x: x['ver-legt']/x['Tier-bestand'],axis=1)
df['DOA'] = df['Datum'].map(lambda x: (x-dt.datetime(2019,7,9)).days)
df['WOA'] = df['DOA'].map(lambda x: math.ceil(x/7))
#remove pen 1 as there is different breeding line
df = df[df['Pen']!='1']
df.to_csv(os.path.join(path_,'egg_summary.csv'), index=False, sep=';')
print(df.shape)
df.head(3)
```

# #Eggs

```python
for v in ['#eggPerTier', '#eggOutsideBoxPerTier']:
    #DOA / pen
    #0_4_E219: not using it: high picks at the beginigna s they dont take everyday, but every few days and hence the sum is higher!
    #avg egg weight/pen
    plt.figure(figsize=(12,6))
    ax = plt.plot()
    df[v] = df[v].astype(float)
    sns.lineplot(x="DOA", y=v, hue='Pen', data=df, 
                 palette=sns.color_palette("RdBu", len(df['Pen'].unique())))
    plt.title(v)
    plt.savefig(os.path.join(path_,v+'_DOA.png'),dpi=300,format='png',bbox_inches='tight')
    plt.show();

    #all WOA
    plt.figure(figsize=(18,8))
    ax = sns.boxplot(x="WOA", y=v, data=df, palette="Set3")
    plt.title(v)
    plt.savefig(os.path.join(path_,v+'_WOA.png'),dpi=300,format='png',bbox_inches='tight')

    #first WOA / pen
    plt.figure(figsize=(18,8))
    ax = sns.boxplot(x="WOA", y=v, hue='Pen', data=df[df['WOA']<=20], 
                     palette=sns.color_palette("RdBu", 20))
    plt.title(v)
    plt.savefig(os.path.join(path_,v+'_FirstWOA.png'),dpi=300,format='png',bbox_inches='tight')
    plt.show()

    #end WOA / pen
    plt.figure(figsize=(18,8))
    ax = sns.boxplot(x="WOA", y=v, hue='Pen', data=df[df['WOA']>52], 
                     palette=sns.color_palette("RdBu", 20))
    plt.title(v)
    plt.savefig(os.path.join(path_,v+'_LastWOA.png'),dpi=300,format='png',bbox_inches='tight')
    #--> if we see variation at the beginign, we might even see mroe variation at the end!! so take a lot of egg data at the end 
```

## Egg weight/pen 

```python
li_df = []
for d in [str(i)+'_4_E219' for i in range(1,12)]:
    print(d)
    xls = pd.ExcelFile(glob.glob(os.path.join(path_,d+'.xls*'))[0])
    df_month = pd.read_excel(xls, 'Eig', header=None)
    df_month = df_month.iloc[3:,]
    df_month.columns = df_month.iloc[0]
    #Note: Tara: number of paper box used
    df_month.rename(columns={'saub. Knick':'PropreFele','Schm. Knick':'DirtyFele','Schm. ganz':'Dirty','Schalen':'ShelDeformation',
                            'Ge-wicht':'AverageEggWeight',1:'Pen','Gesamt-Total':'NumberEggUsedInClassification',
                            'Total Neben.':'TotalBadEggs','Total Klass.':'TotalGoodEgg'}, 
              inplace=True)
    df_month = df_month.drop(3, axis=0)
    #keep only the first table
    df_month = df_month.loc[0:23,]
    df_month['LayingPhase'] = int(d.split('_')[0])
    if df_month.shape[0]!=20:
        print('ERROR: ', d, v,' have wrong shape', df.shape)
    if df_month.shape[1]!=22:
        print('ERROR: ', d, v,' have wrong shape', df.shape)
    li_df.append(df_month)
df_month = pd.concat(li_df)
print(df_month.shape)
df_month.head(3)
```

```python
#avg egg weight/pen
plt.figure(figsize=(12,6))
ax = plt.plot()
df_month.AverageEggWeight = df_month.AverageEggWeight.astype(float)
sns.lineplot(x="LayingPhase", y='AverageEggWeight', hue='Pen', data=df_month, 
             palette=sns.color_palette("RdBu", len(df_month['Pen'].unique())))
plt.title('AverageEggWeight over time per pen')
plt.savefig(os.path.join(path_,'AverageEggWeight_LayingPhase.png'),dpi=300,format='png',bbox_inches='tight')
plt.show();
```

## Egg Classification

```python
#class/pen
for v in ['-47', '47-50', '50-53', '53-65', '65-70', '70-80', '체ber 80', 'TotalBadEggs', 
          'PropreFele', 'DirtyFele', 'Dirty','ShelDeformation']:
    plt.figure(figsize=(12,6))
    ax = plt.plot()
    df_month[v].fillna(0,inplace=True)
    df_month[v+'Ratio'] = df_month.apply(lambda x: x[v]/x['NumberEggUsedInClassification'], axis=1)
    df_month[v+'Ratio'] = df_month[v+'Ratio'].astype(float)
    sns.lineplot(x="LayingPhase", y=v+'Ratio', hue='Pen', data=df_month, 
                 palette=sns.color_palette("RdBu", len(df_month['Pen'].unique())))
    plt.title(v+' Ratio over time per pen')
    plt.savefig(os.path.join(path_,v+'Ratio_LayingPhase_perPen.png'),dpi=300,format='png',bbox_inches='tight')
    plt.show();
```

```python
#class summing all pens
#'-47', '47-50', '50-53', '53-65', '65-70', '70-80', '체ber 80',: not interesting as we have total weight which is better I guess
for v in ['-47', '47-50', '50-53', '53-65', '65-70', '70-80', '체ber 80', 'TotalBadEggs']:
    plt.figure(figsize=(12,6))
    ax = plt.plot()
    df_month[v].fillna(0,inplace=True)
    df_month[v+'Ratio'] = df_month.apply(lambda x: x[v]/x['NumberEggUsedInClassification'], axis=1)
    df_month[v+'Ratio'] = df_month[v+'Ratio'].astype(float)
    df_month[v+'Ratio'] = df_month[v+'Ratio'].fillna(0)
    sns.lineplot(x="LayingPhase", y=v+'Ratio', data=df_month)
    plt.title(v+' Ratio over time per pen')
    plt.savefig(os.path.join(path_,v+'Ratio_LayingPhase.png'),dpi=300,format='png',bbox_inches='tight')
    plt.show();
    
#avg egg weight/pen
plt.figure(figsize=(12,6))
ax = plt.plot()
for v in ['PropreFele', 'DirtyFele', 'Dirty','ShelDeformation']:
    df_month[v].fillna(0,inplace=True)
    df_month[v+'Ratio'] = df_month.apply(lambda x: x[v]/x['NumberEggUsedInClassification'], axis=1)
    df_month[v+'Ratio'] = df_month[v+'Ratio'].astype(float)
df_plt = df_month[['PropreFele'+'Ratio', 'DirtyFele'+'Ratio', 'Dirty'+'Ratio','ShelDeformation'+'Ratio',
                   'LayingPhase']].melt('LayingPhase', var_name='class', value_name='Ratio')
display(df_plt.head())
sns.lineplot(x="LayingPhase", y='Ratio', hue='class', data=df_plt)
plt.title('class Ratio over time')
plt.savefig(os.path.join(path_,'class ratio_LayingPhase.png'),dpi=300,format='png',bbox_inches='tight')
plt.show();
```

```python
#average per pen
li_var = [v+'Ratio' for v in ['-47', '47-50', '50-53', '53-65', '65-70', '70-80', '체ber 80','TotalBadEggs']]
df_plt = df_month.groupby('LayingPhase')[li_var].mean()
display(df_plt.head(2))
ax = df_plt[li_var].plot(kind="bar", stacked=True, figsize=(20,6))
fig = ax.get_figure()
plt.savefig(os.path.join(path_,'Class_Stackedbar.png'),dpi=300,format='png',bbox_inches='tight')
plt.show();
```
