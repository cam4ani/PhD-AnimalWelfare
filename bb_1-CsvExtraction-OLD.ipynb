#### bining with different period

```python
#min and max date
df_ts = pd.read_csv(os.path.join(path_extracted_data, id_run+'_TimeSeries_.csv'), sep=';', parse_dates=['Timestamp', 'day'])
mi = min(df_ts['Timestamp'].tolist())
ma = max(df_ts['Timestamp'].tolist())

li_path = glob.glob(os.path.join(path_extracted_data, 'HensTimeSeries', 
                                 id_run+'_ts_MostFrequentZone_period*_'+str(mi).split(' ')[0]+\
                                 '_'+str(ma).split(' ')[0]+'_*.csv'))
dico_period_lidf = {}
df_bining = pd.DataFrame()
for path_ in li_path:
    if 'VERIFICATION' not in path_:
        h = 'hen_'+path_.split('hen_')[1][:-4]
        period = int(path_.split('period')[1].split('_')[0])
        if period not in dico_period_lidf:
            dico_period_lidf[period] = []
        df_ = pd.read_csv(path_, sep=';', parse_dates=['New_Timestamp'])
        df_['HenID'] = h
        df_.rename(columns={'New_Timestamp':'New_Timestamp_'+str(period),
                            'most_frequent_zone':'most_frequent_zone_'+str(period)}, inplace=True)
        li_info = ['New_Timestamp_'+str(period),'most_frequent_zone_'+str(period),'HenID']
        dico_period_lidf[period].append(df_.filter(li_info))
```

```python
df_all = df_all.sort_values(['Timestamp']) #must be sorted for merge_asof
df_all = df_all.reset_index(drop=True)
for period,li_df in dico_period_lidf.items():
    df_bining = pd.concat(li_df)
    df_bining = df_bining.sort_values(['New_Timestamp_'+str(period)]) #must be sorted for merge_asof
    df_all = pd.merge_asof(df_all, df_bining, by='HenID', left_on=['Timestamp'], right_on=['New_Timestamp_'+str(period)],
                           direction='forward')
df_all = df_all.sort_values(['HenID','Timestamp'])
df_all = df_all.reset_index(drop=True)
df_all.to_csv(os.path.join(path_extracted_data ,id_run+'_records_with_some_cleaned_info_AGGREGATED_'+str(mi).split(' ')[0]+\
                                         '_'+str(ma).split(' ')[0]+'.csv'), sep=';', index=False)
df_all
```

### check the possible sequence of flickering type 2 on the hole cleaning time

```python
df_ = pd.read_csv('D:\\vm_exchange\\AVIFORUM\\data\\extracted_info_breeder_broiler\\WithMicroseconds_records_with_some_cleaned_info_3.csv', sep=';') #parse_dates
df_['flickering2_name'] = df_['flickering2_name'].fillna('')
li_df = []
for i, df_hen in tqdm.tqdm(df_.groupby(['HenID'])):
    df_hen['previous_zone_flickering2'] = [None]+df_hen['zone_flickering2'].tolist()[0:-1]
    df_hen['is_of_interest'] = df_hen.apply(lambda x: (x['flickering2_name']!='') & \
                                            (x['previous_zone_flickering2']=='flickering2'), axis=1)
    df_hen['interesting_sequ'] = df_hen.apply(lambda x: [i for i in df_hen[df_hen['Timestamp']<x['Timestamp']]['flickering2_name'].tolist() if i!=''][-1] if x['is_of_interest'] else None, 
                                       axis=1)
    li_df.append(df_hen)
#put again in one dataframe
df_ = pd.concat(li_df)

df__ = df_[(df_['is_of_interest'])&(df_['interesting_sequ']!=df_['flickering2_name'])].copy()
df__['tuple_sequ'] = df_.apply(lambda x: (x['interesting_sequ'],x['flickering2_name']), axis=1)
#for i in df__.index:
#    display(df_.iloc[i-5:i+5])
#    sys.exit()
df__ = df__.groupby(['tuple_sequ'])['HenID','PenID','Timestamp','log_file_name'].agg(lambda x: tuple(x)).reset_index()
df__['number of cases'] = df__['HenID'].map(lambda x: len(x))
df__ = df__.sort_values(['number of cases'], ascending=False)

df__.to_csv(os.path.join(path_extracted_data, id_run+'_consecutives_flickering2event_info.csv'),
                       index=False, sep=';')
df__
```

```python
#plot
for V in ['HenID','PenID']:
    for type_ in df__['tuple_sequ'].unique():
        c = Counter(df__[df__['tuple_sequ']==type_][V].values[0])
        l = [(k,v) for k,v in c.items()]
        l = sorted(l,key=itemgetter(1),reverse=True)
        fig, ax = plt.subplots(figsize=(10,5))
        width=0.8
        x=[i[0] for i in l]; y=[i[1] for i in l]
        ax.bar(range(len(x)), y, width=width)
        ax.set_xticks(np.arange(len(x)) + width/2)
        ax.set_xticklabels(x, rotation=90, size=8)
        plt.title(type_)
        plt.savefig(os.path.join(path_extracted_data, 'visual', 
                                 id_run+'_consecutives_flickering2event_info_'+V+str(type_)+'.png'), dpi=300,
                                 format='png', bbox_inches='tight')
        plt.show()
```

# Verify lost info depending on zones when bining (normal?)

```python
li_path = glob.glob(os.path.join(path_extracted_data, 'HensTimeSeries', id_run+'_ts_MostFrequentZone_period*.csv'))

dico_mi_li_df = {}
for path_ in li_path:
    h = 'hen_'+path_.split('hen_')[1][:-4]
    if 'VERIFICATION' not in path_:
        period = int(path_.split('period')[1].split('_')[0])
        df_ = pd.read_csv(path_, sep=';', parse_dates=['New_Timestamp'])
        d1 = path_.split('period')[1].split('_')[1]
        d2 = path_.split('period')[1].split('_')[2]
        mi=dt.datetime(int(d1[0:4]),int(d1[5:7]),int(d1[8:]),0,0,0)
        ma=dt.datetime(int(d2[0:4]),int(d2[5:7]),int(d2[8:]),23,59,59)
        n = str(mi)+'   '+str(ma)
        df_['HenID'] = h
        #if 'nbr_lost_duration_per_zone' in df_.columns:
        df_['nbr_lost_duration_per_zone'] = df_['nbr_lost_duration_per_zone'].map(lambda x: eval(x))
        dico_ = dict(functools.reduce(operator.add, map(collections.Counter, df_['nbr_lost_duration_per_zone'].tolist()))) 
        df_['nbr_duration_per_zone'] = df_['nbr_duration_per_zone'].map(lambda x: eval(x))
        dico_all = dict(functools.reduce(operator.add, map(collections.Counter, df_['nbr_duration_per_zone'].tolist()))) 
        if n not in dico_mi_li_df:
            dico_mi_li_df[n] = []
        for z in dico_all.keys():
            dico_mi_li_df[n].append({'period':period, 'HenID':h, 
                                     'percentage_lost_duration':dico_.get(z,0)/dico_all[z]*100, 'zone': z})

#plot
for n, li_df in dico_mi_li_df.items():
    df_plot = pd.DataFrame(li_df)
    #plot for each datarange a plot showing the evolution of duration lost per zone and period
    fig, ax = plt.subplots(figsize=(12,6))
    ax = sns.boxplot(x="period", y="percentage_lost_duration", hue="zone", data=df_plot) #, palette="Set3")
    plt.title(n)
    plt.savefig(os.path.join(path_extracted_data,'visual','for_verification', 
                             id_run+'_percentage_lost_duration_during_bining'+n.split(':')[0][:-3]+'_'+n.split(':')[2][5:-3]+'.png'), dpi=300,
                             format='png', bbox_inches='tight')
    plt.show()

#-->zone 1: si flickering pas possible peut-être que nous devrons le garder en plus deu bining, puis choisir un bining 
#plus petit ou égale à 5mn peut etre :) 
```
