# Accelerometer mvt

```python
df_device = openDevice(config)
df_device = df_device[df_device['sort']=='Tag']
df_device['is_>=20h-<2'] = df_device['Timestamp'].map(lambda x: (x.hour<2) | (x.hour>=20))
df_device['is_day'] = df_device['Timestamp'].map(lambda x: is_day(x, dico_night_hour))
df_device = df_device.sort_values(['Timestamp'], ascending=True)
df_device.head(3)
```

```python
df_ALL = device_Variables(config, df_device)
print(df_ALL.shape)
display(df_ALL.head(3))
display(df_ALL.tail(3))
```

```python
#plot the mvt over entire 24h 
df_device = df_device.sort_values(['Timestamp'], ascending=True)
k = 0
for d,df_ in df_device.groupby(['date']):
    if df_.shape[0]>24*6*100:
        df_['sender'] = df_['sender'].astype('category')
        sns.lineplot(data=df_[(df_['sort']=='Tag')&(df_['sender'].isin(['181','4','5']))], 
                     x='Timestamp', y='MovementCounter', hue='sender', legend='full');
        plt.title(d)
        plt.show()
        k = k+1
    if k==5:
        break
```

```python
#plot the mvt over night per chicken
df_device_ = df_device[~df_device['is_day']].copy()
df_device_['night_level'] = df_device_['Timestamp'].map(lambda x: str(x)[0:-9]+'_'+str(x+dt.timedelta(days=1))[8:10] if\
                                        name_level(x,dico_night_hour) else str(x-dt.timedelta(days=1))[0:-9]+'_'+str(x)[8:10])
display(df_device_[['Timestamp','night_level']].head(3))
df_device_ = df_device_.sort_values(['Timestamp'], ascending=True)
k = 0
for d,df_ in df_device_.groupby(['night_level']):
    if df_.shape[0]>24*6*100:
        df_['sender'] = df_['sender'].astype('category')
        sns.lineplot(data=df_[(~df_['is_day'])&(df_['sort']=='Tag')&(df_['sender'].isin(['181','183','69','4','5']))], 
                     x='Timestamp', y='MovementCounter', hue='sender', legend='full');
        plt.title(d)
        plt.show()
        k = k+1
    if k==5:
        break
        
#--> rarely = 0 even during night (they all seemed to move during night, so would be ard to compute precise variables out of 
#this, we will use logs instead
```

```python
#plot the temperatures over night per chicken. We wont be able to use the temperature until that level of detailed, as no 
#reliability test is planed to verify this
#first add night levelnight
df_device['is_day'] = df_device['Timestamp'].map(lambda x: is_day(x, dico_night_hour))
df_device_ = df_device[~df_device['is_day']].copy()
df_device_['night_level'] = df_device_['Timestamp'].map(lambda x: str(x)[0:-9]+'_'+str(x+dt.timedelta(days=1))[8:10] if\
                                        name_level(x,dico_night_hour) else str(x-dt.timedelta(days=1))[0:-9]+'_'+str(x)[8:10])
display(df_device_[['Timestamp','night_level']].head(3))
#then plot
df_device_ = df_device_.sort_values(['Timestamp'], ascending=True)
k = 0
for d,df_ in df_device_.groupby(['night_level']):
    if df_.shape[0]>24*6*100:
        df_['sender'] = df_['sender'].astype('category')
        sns.lineplot(data=df_[(df_['is_>=20h-<2'])&(df_['sort']=='Tag')&(df_['sender'].isin(['181','183','69','4','5']))], 
                     x='Timestamp', y='Temperature', hue='sender', legend='full');
        plt.title(d)
        plt.show()
        k = k+1
    if k==5:
        break
```
